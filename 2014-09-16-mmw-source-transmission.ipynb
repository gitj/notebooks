```python
def plot_response(filename,ref=None,alt_legend=None,plot_db=True):
    d = np.load(filename)
    fr = d['mmw_freq']
    power = d['power_watts']
    if ref is not None:
        if not np.isscalar(ref):
            ref = np.interp(fr,ref['mmw_freq'],ref['power_watts'])
    else:
        ref = 1e-3
    dbm = 10*np.log10(power/ref)
    if alt_legend is not None:
        label = alt_legend
    else:
        try:
            label = d['comments'][()]
        except:
            label = ''
        label += ('\n'+filename)

    if len(label)-len(filename) > 60:
        label_parts = label.split()
        num_parts = len(label_parts)
        label = (' '.join(label_parts[:num_parts//2])
                 + '\n'
                 + ' '.join(label_parts[num_parts//2:]))
    if plot_db:
        plot(fr/1e9,dbm,label=label)
    else:
        plot(fr/1e9,power/ref,label=label)
            
```
