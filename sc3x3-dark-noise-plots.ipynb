```python
dd = np.array([(x.start_temp,x.f0,x.Q_i,x.atten,x.index,k) for (k,x) in enumerate(nms)])
```

```python
msk = (dd[:,0] < 0.145) & (dd[:,4]==4) & (dd[:,0] >0.135)
```

```python
figure(figsize=(8,6))
fref = nm.f0
for idx in dd[:,-1][msk]:
    nm = nms[int(idx)]
#    plot(nm.fr-fref,abs(nm.s21)/abs(nm.s21).max(),'.-')
#    xlim(-0.005,0.005)
    loglog(nm.fr_coarse,nm.pii_coarse/(nm.f0*1e6)**2,label=('%.1f dBm' % (-52-nm.atten)))
legend(loc='upper right',prop=dict(size='small'))
ylim(1e-19,1e-16)
grid()
xlabel('Hz')
ylabel('1/Hz')
```

```python
figure(figsize=(8,6))
fref = nm.f0
for idx in dd[:,-1][msk]:
    nm = nms[int(idx)+4]
#    plot(nm.fr-fref,abs(nm.s21)/abs(nm.s21).max(),'.-')
#    xlim(-0.005,0.005)
    l, =  loglog(nm.fr_fine[:70],nm.pii_fine[:70]/(nm.f0*1e6)**2,label=('%.1f dBm' % (-52-nm.atten)))
    loglog(nm.fr_coarse,nm.pii_coarse/(nm.f0*1e6)**2,color=l.get_color(),lw=2)
legend(loc='upper right',prop=dict(size='small'))
ylim(1e-19,1e-16)
grid()
xlabel('Hz')
ylabel('1/Hz')
```
