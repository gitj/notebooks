```python
nm = nmons[0]
nm.atten
```

```python
device = []
amp = []
for k in range(13):
    proj = nm.resonator_model.project_s21_to_delta_freq(meas_freqs[k],s21s[k])
    pxx,fr = mlab.psd(proj.real/nm.resonator_model.f_0/1e6,NFFT=512,Fs=rnc.sweep_20141022124921.timestream_group.sample_rate[0])
    loglog(fr,pxx,color=plt.cm.hsv(k/13.))
    device.append(pxx[(fr>80) & (fr<150)].mean())
    amp.append(pxx[fr>4e3].mean())
#xscale('log')
figure()
semilogy(meas_freqs,device,'o-')
semilogy(meas_freqs,amp,'x-')
figure()
semilogy(meas_freqs,np.array(device)-np.array(amp))
```
