```python
def dxdP(Po,taumax=100e-6,nstar=400,alpha=0.66,N0=1.72e10,Delta=1.76*1.38e-23*1.46,Vl=1870,nu=150e9,eta=1,etapb=0.7,
         S2=3.8):
    a1 = alpha*S2*taumax*etapb*eta/(2*N0*Delta*6.626e-34*nu*Vl/1.6e-19)
    a2 = np.sqrt(1+4*etapb*eta*Po*taumax/(6.626e-34*nu*Vl*nstar))
    return a1/a2
```
