```python
fig,ax = plt.subplots(1,1,figsize=(3,2))
ld = lt[(lt.atten==32.) & (lt.timestream_modulation_duty_cycle==0.0)]
hd = ht[(ht.atten==32.) & (ht.timestream_modulation_duty_cycle==0.0)]
colors = cm.spectral_r(np.linspace(0,1,20))
for k,blah in enumerate(axs.flatten()):
    if k not in [1,3,13,15]:
        continue
    lk = ld[ld.resonator_id==k]
    hk = hd[hd.resonator_id==k]
    maxf = cal[cal.resonator_id==k].zbd_power_f_r_max.iloc[0]
    thiscal = cal[cal.resonator_id==k]
    #print k,maxf
    maxf = lk.f_0.max()
    lx = (maxf-lk.f_0)/maxf
    hx = (maxf-hk.f_0)/maxf
    hpow = kid_readout.analysis.dataframe.X_inverse(hx,thiscal.zbd_power_XI_fit_P_0.iloc[0], thiscal.zbd_power_XI_fit_P_star.iloc[0],
                                                    thiscal.zbd_power_XI_fit_X_0.iloc[0])
    lpow = kid_readout.analysis.dataframe.X_inverse(lx,thiscal.zbd_power_XI_fit_P_0.iloc[0], thiscal.zbd_power_XI_fit_P_star.iloc[0],
                                                    thiscal.zbd_power_XI_fit_X_0.iloc[0])
    hpow = thiscal.zbd_power_XI_fit_P_0.iloc[0] + hpow
    lpow = thiscal.zbd_power_XI_fit_P_0.iloc[0] + lpow
    hpow = hpow*1e-6*1e12
    lpow = lpow*1e-6*1e12
    allpow = np.concatenate((lpow,hpow))
    alltemp = np.concatenate((lk.sweep_primary_load_temperature, hk.sweep_primary_load_temperature))
    order = alltemp.argsort()
    alltemp = alltemp[order]
    allpow = allpow[order]
    #ax.plot(lk.sweep_primary_load_temperature,lx,'o',color=colors[k])
    #ax.plot(hk.sweep_primary_load_temperature,hx,'x',color=colors[k],mew=2)
    ax.semilogy(alltemp,allpow,color=colors[k])
    ax.semilogy(hk.sweep_primary_load_temperature,hpow,'o',color=colors[k],ms=4,mew=0)
    ax.semilogy(lk.sweep_primary_load_temperature,lpow,'x',color=colors[k],mew=1.5,ms=4)
    
ax.grid()
ax.set_ylabel('Incident power [pW]')
ax.set_xlabel('Blackbody temperature [K]')
fig.savefig('/home/gjones/paper_plots/2015-03-09-mmw-vs-Tbb.pdf',bbox_inches='tight')
#yscale('linear')

```
