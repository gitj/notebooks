```python
from kid_readout.utils.easync import EasyNetCDF4
```

```python
ll = 625/5
def load_led(ncf):
    nc = EasyNetCDF4(ncf)
    tsns = nc.timestreams.groups.keys()
    ts0 = nc.timestreams.groups[tsns[0]].variables['data'][:].view('complex')
    ts1 = nc.timestreams.groups[tsns[1]].variables['data'][:].view('complex')
    t = np.arange(ts0.shape[1])*(2**14/256e6)
    dar = ts0.T[:ll*(ts0.shape[0]/ll),:].reshape((-1,ll,4))
    nc.group.close()
    return t,ts0,ts1,dar
```

```python
t,ts0,ts1,dar = load_led('/home/data/2013-12-20_154506.nc')
```
