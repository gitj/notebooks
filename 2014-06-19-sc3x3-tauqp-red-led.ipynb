```python
T = np.array([100,150,200,250,270,280,290])
tau = np.array([820,850,890,560,250,200,120])*.45
semilogy(T,tau,'o')
semilogy(Tm*1000.,taum)
ylim(10,1000)
xlim(90,300)
grid()
ylabel(r'$\tau_{qp}$ [$\mu$s]')
xlabel('Base Temperature [mK]')
title('QP lifetime measured at ~890 MHz\nwith LED pulse falling edge\STAR Cryo 3x3\ntau_max ~ 400 $\mu$s',size='small')
```
