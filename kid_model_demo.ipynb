```python
f = np.linspace(99.98e6,100.02e6,1000)
for Q in [1e4,1e5,2e5]:
    p = khalil.create_model(Q=Q,Q_e=1e5)
    s21 = khalil.generic_s21(p,f)
    plot(s21.real,s21.imag,'.-',label=('Q=%.2g' % Q))
legend()
```

```python
figure(figsize=(4,4))
f = np.linspace(99.98e6,100.02e6,1000)
Q_i = 1e5
Q_e = 1e5
for Q_i in [1e4,1e5,5e5]:
    Q = 1/(1/Q_i + 1/Q_e)
    p = khalil.create_model(Q=Q,Q_e=Q_e)
    s21 = khalil.generic_s21(p,f)
    plot(s21.real,s21.imag,'.-',label=('Qc=%.2g, Qi=%.2g, Qr=%.2g' % (Q_e,Q_i,Q)))
legend(prop=dict(size='small'),loc='upper left')
xlim(-1,1)
ylim(-1,1)
grid()
```
