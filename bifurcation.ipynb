```python
data = []
power_levels = []
for name,sweep in nc.sweeps.groups.items()[5:]:
    print name
    try:
        s21 = sweep.variables['s21'][:].view('complex128')
        fr = sweep.variables['frequency'][:]
        epoch = sweep.groups['datablocks'].variables['epoch'][0]
        idx = np.flatnonzero(nc.hw_state.epoch[:]<epoch)[-1]
        dac_atten = nc.hw_state.dac_atten[idx]
        power_level = -90+(12.0-dac_atten)
        frs,s21s,f0s = proc_sweep.extract_fine_sweeps(fr,s21)
        data.append((name,frs,s21s,f0s))
        power_levels.append(power_level)
    except Exception,e:
        print e

order = np.array(power_levels).argsort()
for k in order:
    name,frs,s21s,f0s = data[k]
    power_level = power_levels[k]
    figure(figsize=(10,6))
    plot((frs.T-f0s)[:,:4]*1000,20*np.log10(np.abs(s21s.T))[:,:4],'.-')
    xlim(-10,10)
    xlabel('kHz')
    ylabel('dB')
    grid()
    title("%s %.1f dBm/tone" % (name,power_level))
```
