```python
centers = np.random.rand(7,2)*0.5-1
```

```python
centers
```

```python
msk = np.zeros((1000,1000),dtype=np.bool)
radius = 0.04
for k in range(centers.shape[0]):
    msk = msk | ((X-centers[k,0])**2 + (Y-centers[k,1])**2 < radius**2)
```

```python
Z = np.zeros((1000,1000))
Z[msk] = 1
Z = Z - Z.mean()
```

```python
figure(figsize=(8,8))
imshow(Z)
```

```python
Zf = np.abs(np.fft.fftshift(np.fft.fft2(Z)))
```

```python
figure(figsize=(8,8))
imshow(Zf)
xlim(400,600)
ylim(400,600)
```

```python
figure(figsize=(8,8))
imshow(log(Zf))
xlim(400,600)
ylim(400,600)
```

```python
def simimage(f):
    return np.fft.fftshift(np.abs(np.fft.ifft2(f))**2)
```

```python
noise = np.random.randn(1000,1000) + 1j*np.random.randn(1000,1000)
```

```python
def gauss2d(xwidth,ywidth,theta):
    nx = X.shape[0]
    ny = Y.shape[0]
    fs = xwidth/float(nx)
    ts = ywidth/float(ny)
    c2 = np.cos(theta)**2
    s2 = np.sin(theta)**2
    a = c2/(2*fs**2) + s2/(2*ts**2)
    b = -np.sin(2*theta)/(4*fs**2) + np.sin(2*theta)/(4*ts**2)
    c = s2/(2*fs**2) + c2/(2*ts**2)
    filt = np.exp(-(a*X**2 + 2*b*X*Y + c*Y**2))
    return filt
```

```python
img = simimage(gauss2d(20,50,0.1)*noise)
img = img-img.mean()
```

```python
figure(figsize=(10,10),dpi=100)
imshow(img,cmap=cm.gray)
colorbar()
```

```python
psd = (100*gauss2d(2,1,0)+0.5*gauss2d(20,30,0.0)+0.1*gauss2d(100,2,0.3))*noise + 0.01*noise
img = simimage(psd)
img = img-img.mean()
```
