```python
X,Y = np.meshgrid(np.linspace(0,1,1000),np.linspace(0,1,1000))
```

```python
X.shape
```

```python
Z = np.zeros((1000,1000))
```

```python
centers = np.random.rand(7,2)*0.5
```

```python
centers
```

```python
msk = np.zeros((1000,1000),dtype=np.bool)
for k in range(centers.shape[0]):
    msk = msk | ((X-centers[k,0])**2 + (Y-centers[k,1])**2 < 0.025**2)
```

```python
Z = np.zeros((1000,1000))
Z[msk] = 1
Z = Z - Z.mean()
```

```python
figure(figsize=(8,8))
imshow(Z)
```

```python
Zf = np.abs(np.fft.fftshift(np.fft.fft2(Z)))
```

```python
figure(figsize=(8,8))
imshow(Zf)
xlim(400,600)
ylim(400,600)
```

```python
figure(figsize=(8,8))
imshow(log(Zf))
xlim(400,600)
ylim(400,600)
```
