# Reading out multiple tones

The system can currently read out 4 tones at a time using the basic software. The new UDP Catcher software will allow more simultaneous tones (currently ~16, but ~128 after more optimization).

Loading stimulus tones takes about 1 second for $2^{20}$ samples, so scanning in frequency by loading a new waveform is very inefficient. Instead we'll typically load a waveform with many stimulus tones, and then read out $n$ at a time, where $n$ will be 4 today.

So now we can try to read out all of the bins, 4 at a time:

```python
tones = []
data = []
for k in range(len(ri.tone_bins)/4):
    ri.select_fft_bins(range(k*4,(k+1)*4))
    time.sleep(0.1) #sleep for a moment for the data to become fresh. could do this smarter, but this works
    d,addr = ri.get_data(2)
    data.append(d)
    tones.append(ri.tone_bins[ri.readout_selection])
data = np.concatenate(data,axis=1)  #stack all the data into one numpy array with axes time,channel
tones = np.concatenate(tones) #stack the tones into one long array
davg = data.mean(0) #compute the average of the data along the time axis for convenience
```

```python
freqs = ri.fs*tones/ri.tone_nsamp  # convert the tone locations to frequencies
```

```python
plot(freqs,20*np.log10(np.abs(davg)))  # plot the transmission in dB power
xlabel('MHz')
ylabel('dB')
grid()
```

You might have noticed the dips in the ADC noise floor earlier on. Now we can see them more clearly. Based on the separation of the two dips (near 30 and 150 MHz), can you determine the length of the transmission line resonator I have connected in the circuit?

You can also see a little "scruff" on the plot. We'll investigate that more next.

First, play with the data a bit. Here's a plot of the power spectra of each FFT bin "in context"

```python
for k,fr in enumerate(freqs):
    pxx,f = mlab.psd(data[:,k],NFFT=256,Fs=ri.fs/ri.nfft)
    plot(f+fr,10*np.log10(pxx))
```

In principle, we have the phase info for this data too, so we can plot the real and imaginary parts etc.

```python
plot(freqs,davg.real)
plot(freqs,davg.imag)
```

```python
plot(freqs,np.angle(davg))
```
