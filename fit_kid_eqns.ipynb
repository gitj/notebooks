```python
ridxs = [x for x in list(set(jpl.ridx)) if np.isfinite(x)]
for k in ridxs:
    figure()
    subplot(2,1,1)
    #r0 = scd[(scd.ridx ==k) & (scd.atten == 15)]
    r0 = jpl[(jpl.ridx ==k) & (jpl.atten == 37)]
    r0 = r0.sort('init_temp')
    T = r0.init_temp
    Qi = r0.Q_i
    f0 = r0.f_0
    Qi_err = r0.Q_err
    f0_err = r0.f_0_err
    km = kid_readout.analysis.kid_eqns.KIDModel(f0_nom=r0.f_0.min()*1e6, cap=r0.Cg, Lg=r0.Lg,delta_0=1e-10,F_TLS=1.0,Tc=1.46,nqp0=1,
                                                delta_loss=0, sigman=1e5,Tbase=0.3)
    #km.params['delta_loss'].vary = False
    km.params['T_star'].vary = False
    #km.params['Tc'].vary = False
    #km.params['nqp0'].vary = False
    km.params['F_TLS'].vary = False
    km.params['delta_0'].vary = False
    #km.params['sigman'].max = 1.1e5
    
    km.fit_f0_qi(T,f0*1e6,Qi,f0_err*1e6,Qi_err)
    #km.fit_qi(T,Qi,Qi_err)
    #plot(T,Qi,'.-')
    errorbar(T,Qi,Qi_err)
    plot(T,km.total_Qi(T),lw=3)
    grid()
    ylabel('$Q_i$')
    pstr = '\n'.join([('%s: %.4g +- %.4g' % (x.name,x.value,x.stderr)) for x in km.params.values() if x.vary == True])
    ax = gca()
    ax.text(0.02,0.04,pstr,fontdict=dict(size='xx-small'),transform=ax.transAxes,bbox=dict(facecolor='w'))

    #plot(T,km.Qi(T),lw=3)
    #plot(T,r0.Q_e_real/10)
    subplot(2,1,2)
    errorbar(T,f0*1e6,f0_err*1e6, label='data')
    plot(T,km.f_res(T)*(1+km.tls_shift(T)), label='model')
    xlabel('T [K]')
    ylabel('$f_0$ [Hz]')
    grid()
    legend(loc='lower left')
    lmfit.report_fit(km.params)
```
