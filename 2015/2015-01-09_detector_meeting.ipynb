# Response to millimeter-wave source

Fit the following to resonator parameters such as $x$ or $Q_i^{-1}$:
$$
z = z_0 \left( (1 + (P + P_0) / P_*)^{1/2} -1 \right),
$$
where $z_0$ sets the scale of $z$, $P$ is the known applied power, $P_0$ is a power offset, and $P_*$ is the characteristic total power for $z$ to respond. Then,
$$
\frac{d z}{dP}
  =
  \frac{z_0}{2 P_*} (1 + (P + P_0) / P_*)^{-1/2}.
$$

```python
def z(params, P):
    z_0 = params['z_0'].value
    P_0 = params['P_0'].value
    P_star = params['P_star'].value
    return z_0 * ((1 + (P + P_0) / P_star)**(1/2) - 1)

def dz_dP(params, P):
    z_0 = params['z_0'].value
    P_0 = params['P_0'].value
    P_star = params['P_star'].value
    return z_0 / (2 * P_star) * (1 + (P + P_0) / P_star)**(-1/2)

def residual(params, P, data, errors=1):
    return (data - z(params, P)) / errors

def fit_z(P, data, initial, errors=1):
    return lmfit.minimize(residual, initial, args=(P, data, errors))
    
def guess(z_0=1, P_0=1, P_star=1):
    params = lmfit.Parameters()
    params.add('z_0', value=z_0)
    params.add('P_0', value=P_0, min=0)
    params.add('P_star', value=P_star, min=0)
    return params
```

This is the inverse of $z$, which is useful for reconstructing power from resonator parameters:
$$
P(z) = P_* (1 + z / z_0)^2 - 1) - P_0.
$$

```python
def z_inverse(params, z):
    z_0 = params['z_0'].value
    P_0 = params['P_0'].value
    P_star = params['P_star'].value
    return P_star * ((1 + z / z_0)**2 - 1) - P_0
```

The spectra plotted above are
$$
\mathrm{NEP}_{\mathrm{ZBD}}^2
  =
  \frac{S_{xx}}{(dx / dP)^2},
$$

The limiting value of the quantity $\mathrm{NEP} / P$ is $(2 / \Delta \nu)^{1/2}$:

```python
bandwidth = 2 / (1.8e-5)**2
1e-9 * bandwidth  # in GHz
```

```python
from scipy.constants import h
nu = 155e9
#bandwidth = 13e9

coupler_20_dB = 1e-2  # Nominal value; probably solid
waveguide_loss = 1e-1  # Handwaving
eccosorb_transmission = 0.265  # Brad's calculation for 0.25"; solid
dilution = 4e-3  # Pure handwaving: need to do GRASP; this will vary between detectors.
zbd_to_detector = coupler_20_dB * waveguide_loss * eccosorb_transmission * dilution
print(zbd_to_detector, eta * zbd_to_detector)

# The optical efficiency from sims
eta = 0.7

PP = 1e-6 * np.logspace(-4, 2, 1e3)  # in watts
P_incident = zbd_to_detector * PP
```

```python
shot_noise_2 = (2 * h * nu * P_incident)
wave_noise_2 = (2 * P_incident**2 / bandwidth)
constant_2 = np.ones_like(P_incident) * 6e-17**2
```

```python
plt.loglog(1e6 * PP, wave_noise_2**(1/2), '--b', label='wave')
plt.loglog(1e6 * PP, shot_noise_2**(1/2), '--g', label='shot')
plt.loglog(1e6 * PP, constant_2**(1/2), '--k', label='constant')
plt.loglog(1e6 * PP, (shot_noise_2 + wave_noise_2 + constant_2)**(1/2), '-r', label='sum')

atten = 32
mask = ((source.atten==atten) &
        (source.timestream_modulation_duty_cycle==0))

for channel, group in source[mask].groupby('channel'):
    group.sort('zbd_power', inplace=True)
#    plt.loglog(1e12 * zbd_to_detector * group.zbd_power,
#               zbd_to_detector * (group.NEP2_zbd_fit_device + group.NEP2_zbd_fit_amplifier)**(1/2), '-k', alpha=0.3)
    plt.loglog(1e12 * zbd_to_detector * group.zbd_power,
               zbd_to_detector * group.NEP2_zbd_fit_device**(1/2), '-k', alpha=0.3)

plt.ylim(1e-17, 1e-15)
plt.xlabel('incident power [pW]')
plt.ylabel('$\mathrm{NEP} \; [\mathrm{W} \, \mathrm{Hz}^{-1/2}$]')
plt.legend(loc='best')

plt.rcParams['figure.facecolor'] = 'gray'
plt.rcParams['figure.subplot.bottom'] = 0.18
#plt.savefig("NEP_work-in-progress.pdf")
```
