```python
ys = np.arange(-20,20)*np.sqrt(3)/2.
```

```python
coords = []
for row,y in enumerate(ys):
    if row % 2 ==0:
        coords.extend(zip(np.arange(-20,20),y*np.ones((40,))))
    else:
        coords.extend(zip(np.arange(-20,20)+0.5,y*np.ones((40,))))
coords = np.array(coords)
```

```python
mask = np.sqrt(coords[:,0]**2+coords[:,1]**2)<5
```

```python
plot(coords[mask,0],coords[mask,1],'o')
```

```python
def hex_coords(nrings):
    if nrings ==1:
        return np.array([(0,0)])
    nrow_pairs = nrings-1
    this_row = nrings*2-1
    points = [(k,0) for k in range(-(nrings-1),this_row-nrings+1)]
    for row in range(1,nrow_pairs+1):
        this_row = this_row - 1
        start = -(nrings-1)
        row_points = [(k+row*0.5,row*np.sqrt(3)*sign/2) for sign in (-1,1) for k in range(start,start+this_row)]
        points.extend(row_points)
    return np.array(points)
    
    
```

```python
hc = hex_coords(11)
#plot(hc[:,0]*4.8,hc[:,1]*4.8,'o')
ax = gca()
for xy in hc:
    p = Circle(xy*4.8,4.8/2,ec='none')
    ax.add_patch(p)
p  = Circle((0,0),50,fc='none')
ax.add_patch(p)
print len(hc)
axis('equal')
```
