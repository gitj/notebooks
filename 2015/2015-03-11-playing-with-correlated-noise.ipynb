```python
ns = ts - ts.mean()
ns2 = ts2 - ts2.mean()
ns3 = ts3 - ts3.mean()
```

```python
pxx,fr = mlab.psd(ns,NFFT=2**16,Fs=256e6/2**14)
pyy,fr = mlab.psd(ns2,NFFT=2**16,Fs=256e6/2**14)
pxy,fr = mlab.csd(ns,ns2,NFFT=2**16,Fs=256e6/2**14)
```

```python
loglog(fr,abs(pxy))
loglog(fr,abs(pxx))
loglog(fr,pxx-abs(pxy)**2/pyy)
```

```python
semilogx(fr,abs(pxy)/np.sqrt(pxx*pyy))
```
