```python
%pylab inline
import kid_readout.analysis.archive
```

```python
import kid_readout.analysis.noise_fit
```

```python
pkls = kid_readout.analysis.archive.load_noise_pkl('/data/readout/pkl/2015-05-19_101902_mmw_continuous-wave_148_GHz.pkl')
```

```python
len(pkls)
```

```python
snm = pkls[-15-48]
```

```python
blah = snm.plot()
```

```python
snm.mmw_atten_turns
```

```python
fr,S,evals,evects,angles,piq=kid_readout.analysis.iqnoise.pca_noise(snm.fractional_fluctuation_timeseries,Fs=snm.timeseries_sample_rate,use_log_bins=False,use_full_spectral_helper=True)
```

```python
fit = kid_readout.analysis.noise_fit.fit_single_pole_noise(snm.pca_freq,snm.pca_eigvals[1,:])
```

```python
figure(figsize=(18,6))
loglog(fr,evals.T)
loglog(snm.pca_freq,snm.pca_eigvals.T,color='k')
loglog(fit.x_data,fit.model(),'r')
grid()
xlabel('Hz')
ylabel('1/Hz')

figure(figsize=(18,6))
semilogy(fr,evals.T)
semilogy(snm.pca_freq,snm.pca_eigvals.T,color='k')
semilogy(fit.x_data,fit.model(),'r')
xlim(0,1000)
ax = gca()
#for spur in 1.4*np.arange(1,30):#[1.4,39,59,78,116,141,144,156,178,207]:
#    ax.text(spur,1e-17,(' %.1f' % spur))
#    ax.axvline(spur,alpha=0.5)
xlabel('Hz')
ylabel('1/Hz')
ax.axhline(0.9e-19,color='r')
ylim(1e-20,1e-17)

figure(figsize=(18,6))
semilogy(fr,evals.T)
semilogy(snm.pca_freq,snm.pca_eigvals.T,color='k')
semilogy(fit.x_data,fit.model(),'r')
xlim(0,40)
ax = gca()
for k,spur in enumerate(1.4*np.arange(1,25)):#[1.4,39,59,78,116,141,144,156,178,207]:
    ax.text(spur,((k%2)+1)*1e-17,(' %d' % (k+1)))
    ax.axvline(spur,alpha=0.5)
xlabel('Hz')
ylabel('1/Hz')
ax.axhline(0.9e-19,color='r')
ylim(1e-20,2.5e-17)
```
