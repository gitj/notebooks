```python
def add_temps(d):
    package = []
    load = []
    for epoch in d.epoch:
        pt,blah,lt,bbla = kid_readout.utils.starcryo_temps.get_temperatures_at(epoch)
        package.append(pt)
        load.append(lt)
    d['primary_package_temperature'] = package
    d['primary_load_temperature'] = load
    return d
```
