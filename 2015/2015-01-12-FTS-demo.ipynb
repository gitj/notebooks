```python
def getBP(f):
    y = np.zeros(f.shape)
#    y[np.abs(f-200e9).argmin()] = 1
#    y[np.abs(f-201e9).argmin()] = 1
    y[(f>140e9) & (f<160e9)] = 1
    y[(f>280e9) & (f<320e9)] = 0.1
    return y
```

```python
def plotH(dx):
    fig,axs = subplots(2,2,figsize=(8,8))
    f = np.linspace(0,1e12,1000)
    H = 1+np.cos(f*2*np.pi*dx/3e10)
    axs[0,0].plot(f/1e9,H)
    bp = getBP(f)
    axs[0,0].plot(f/1e9,H*bp,'r')
    axs[0,0].set_ylim(-.1,2.1)
    dxs = np.arange(-3,3,xstep)
    cgr = np.array([((1+np.cos(f*2*np.pi*ddx/3e10))*bp).mean() for ddx in dxs])
    axs[1,0].plot(dxs,cgr,'b')
    axs[1,0].plot(dxs[abs(dxs)<dx],cgr[abs(dxs)<dx],'r')
    cg1 = (H*bp).mean()
    axs[1,0].plot([-dx,dx],[cg1,cg1],'ro')
    axs[1,0].set_ylim(0,0.05)
    axs[1,0].set_xlim(-3,3)
    fm.raw_position = dxs[(dxs>=0)]/2
    fm.raw_visibility = np.array(cgr)[(dxs>=0)]
    fm.raw_visibility[(fm.raw_position>=dx/2.)] = (fm.raw_visibility[(fm.raw_position<dx/2.)].mean())
    fm.analyze(0)#,detrend=detrend_none)
    fm.plot_spectrum(ax=axs[0,1])
    fm.plot_interferogram(ax=axs[1,1])
    fm.plot_interferogram(ax=axs[1,1],windowed=True)
    axs[0,1].set_ylim(-50,0)
```
