```python
def fold_data(data,period_in_samples):
    phase = np.round(np.mod(np.arange(data.shape[0])/period_in_samples,1)*period_in_samples).astype('int')
    profile_r = np.bincount(phase,weights=data.real)
    profile_i = np.bincount(phase,weights=data.imag)
    profile = profile_r +1j*profile_i
    hits = np.bincount(phase)
    profile = profile/hits
    return profile,hits
    
```

```python
profile,hits = fold_data(fl,1383.475)
print profile.real.max()
```
