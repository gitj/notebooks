```python
def compute_poynting(E,H,nu=150e9):
    """
    Compute the Poynting vector, using the normalization from Brad's code
    """
    lambd_mm = 1000*3e8/nu
    k = 2*np.pi/lambd_mm
    Pu = k**2 * (E[1] * H[2].conj() - E[2] * H[1].conj()).real / 2
    Pv = k**2 * (E[2] * H[0].conj() - E[0] * H[2].conj()).real / 2
    Pr = k**2 * (E[0] * H[1].conj() - E[1] * H[0].conj()).real / 2
    return Pu,Pv,Pr
def compute_normalized_poynting(E,H,nu=150e9):
    """
    Compute the Poynting vector, normalized to 4pi watts input power
    """
    Pu,Pv,Pr = compute_poynting(E,H,nu=nu)
    pi4 = 4*np.pi
    return (Pu/pi4), (Pv/pi4), (Pr/pi4)
```

```python
Pu1, Pv1, Pr1 = compute_normalized_poynting(E1,H1)
Pu2, Pv2, Pr2 = compute_normalized_poynting(E2,H2)
Pu3, Pv3, Pr3 = compute_normalized_poynting(E3,H3)
```
