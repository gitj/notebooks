```python
def load_fts_data(fname):
    fh = open(fname)
    lines = fh.readlines()
    fh.close()
    note = [x for x in lines if x.find('Note:') >= 0]
    if len(note) < 1:
        print "didn't find note"
        note = ""
    else:
        note=note[0].strip()[5:]
    data = []
    for line in lines:
        if line[0] in ['\t', '\n']:
            continue
        parts = line.split()
        x = float(parts[0])
        y = float(parts[1])
        data.append((x,y))
    return np.array(data)
```
