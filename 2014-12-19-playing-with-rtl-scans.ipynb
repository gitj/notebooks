```python
import time
```

```python
def get_sweep(flo=460,fhi=470,fstep=1,nsamp=2**20):
    for freq in np.arange(flo,fhi,fstep)*1e6:
        r.set_center_freq(freq)
        time.sleep(0.02)
        d = r.read_samples(nsamp)
        d = d[nsamp/32:]
        d = d-d.mean()
        pxx,fr = mlab.psd(d,NFFT=256,Fs=1024e3)
        plot(fr+freq,10*np.log10(pxx))
        dbm = 10*np.log10(pxx)
        peak = dbm.max()
        if peak > -104:
            print "peak at", (fr[dbm.argmax()]+freq), peak,"dBm"
```

```python
figure(figsize=(18,8))
get_sweep(flo=50,fhi=1150,nsamp=2**16)
```
