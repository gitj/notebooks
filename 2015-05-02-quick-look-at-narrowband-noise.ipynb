Plot the resonance sweeps. Since this is unnormalized, the fact that they all lie at the same place on the y axis again implies that the readout power level was the same.
Since the resonances are in the same basic place, the absorbed power is very similar.

```python
fig,axs=plt.subplots(1,2,figsize=(12,6))
fig.subplots_adjust(wspace=.3)
axs[0].plot(nm1.sweep_freqs_MHz,20*np.log10(np.abs(nm1.sweep_s21)),label='narrow1')
axs[0].plot(nm2.sweep_freqs_MHz,20*np.log10(np.abs(nm2.sweep_s21)),label='narrow2')
axs[0].plot(nm3.sweep_freqs_MHz,20*np.log10(np.abs(nm3.sweep_s21)),'.-',label='wide1')
axs[0].plot(nm4.sweep_freqs_MHz,20*np.log10(np.abs(nm4.sweep_s21)),'.-',label='wide2')
#axs[0].plot(nm5.sweep_freqs_MHz,20*np.log10(np.abs(nm5.sweep_s21)),'.-',label='cw')
axs[0].set_xlim(96.06,96.07)
axs[0].legend(loc='lower left',prop=dict(size='small'))
axs[0].set_xlabel('MHz')
axs[0].set_ylabel('dB')

axs[1].loglog(nm1.pca_freq,nm1.pca_eigvals[1,:])
axs[1].loglog(nm2.pca_freq,nm2.pca_eigvals[1,:])
axs[1].loglog(nm3.pca_freq,nm3.pca_eigvals[1,:])
axs[1].loglog(nm4.pca_freq,nm4.pca_eigvals[1,:])
#axs[1].loglog(nm5.pca_freq,nm5.pca_eigvals[1,:])
axs[1].set_xlabel('Hz')
axs[1].set_ylabel('Sxx [1/Hz]')
```

The extra noise is easily visible, even when just looking at the timestream:

```python
t = np.arange(100000)/nm1.timeseries_sample_rate
plot(t,1e6*nm1.fractional_fluctuation_timeseries.real[:100000],label='narrow')
plot(t,1e6*nm3.fractional_fluctuation_timeseries.real[:100000],label='wide')
legend(loc='lower right',prop=dict(size='small'))
ylabel('fractional frequency [ppm]')
xlabel('seconds')
```
