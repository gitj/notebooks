```python
nm = noms[0]
```

```python
zbd_voltage = np.array([nm.zbd_voltage for nm in powsweep])
zbd_power = zbd_voltage/2000
f_0 = np.array([nm.fit_params['f_0'].value for nm in powsweep])

qi = np.array([nm.Q_i for nm in powsweep])
q = np.array([nm.fit_params['Q'].value for nm in powsweep])
frac_f0 =(f_0/f_0.max()-1)
```

```python
pp = np.polyfit(frac_f0,zbd_power,2)
print pp
plot(zbd_power*1e6,1e6*frac_f0,'o',label='data')
y = np.linspace(frac_f0.min(),frac_f0.max(),100)
plot(np.polyval(pp,y)*1e6,1e6*y,label='quadratic fit')
grid()
legend(loc='upper right', prop=dict(size='small'))
xlabel('Measured mm-wave power at source ($\mu$W)')
ylabel('Fractional frequency shift (ppm)')
#ylim(-160,0)
```

```python
df = np.linspace(-140,0,100)*1e-6
dpdf = 2*pp[0]*df+pp[1]
plot(np.polyval(pp,df)*1e6,1/dpdf)
grid()
```

```python
loglog(zbd_power,-frac_f0,'.-')
grid()
xlabel('Measured mm-wave power at source (W)')
ylabel('Fractional frequency shift')
```

```python
plot(zbd_power*1e6,1e6/qi,'o',label='Qi')
plot(zbd_power*1e6,1e6/q,'o',label='Qr')
grid()
legend(loc='upper left', prop=dict(size='small'))
ylabel('$10^6/Q$')
xlabel('Measured mm-wave power at source ($\mu$W)')
```

```python
figure(figsize(12,6))
for k,nm in enumerate(powsweep):
    #print -np.log10((nm.zbd_voltage/2000.0)/36e-6)/5
    if k % 2 == 0 and nm.zbd_voltage > 0:
        pw = nm.zbd_voltage/2000
        if pw > 1e-7:
            label = ('%.1f $\mu$W' % (1e6*pw))
        else:
            label = ('%.1f nW' % (1e9*pw))
    else:
        label = None
    loglog(nm.pca_freq,nm.pca_eigvals[1,:],color = plt.cm.jet(1-(-np.log10((nm.zbd_voltage/2000.0)/36e-6)/5)),
           label=label)

grid()
xlabel('Hz')
xlim(1,4e4)
ylabel('Fractional frequency noise (1/Hz)')
legend(loc='upper right', prop=dict(size='small'))
```

```python
figure(figsize(12,6))
for k,(ff0,nm) in enumerate(zip(frac_f0,powsweep)):
    #print -np.log10((nm.zbd_voltage/2000.0)/36e-6)/5
    if k % 2 == 0 and nm.zbd_voltage > 0:
        pw = nm.zbd_voltage/2000
        if pw > 1e-7:
            label = ('%.1f $\mu$W' % (1e6*pw))
        else:
            label = ('%.1f nW' % (1e9*pw))
    else:
        label = None
    loglog(nm.pca_freq,abs(2*pp[0]*ff0+pp[1])**2*nm.pca_eigvals[1,:],color = plt.cm.jet(1-(-np.log10((nm.zbd_voltage/2000.0)/36e-6)/5)),
           label=label)

grid()
xlabel('Hz')
xlim(1,4e4)
ylabel('Fractional frequency noise (1/Hz)')
legend(loc='upper right', prop=dict(size='small'))
```

```python
device = np.array([nm.pca_eigvals[1,74:76].mean() for nm in powsweep])
amp = np.array([nm.pca_eigvals[1,-10:].mean() for nm in powsweep])
```

```python
loglog(zbd_power,abs(2*pp[0]*frac_f0+pp[1])*np.sqrt(device),'.-')
loglog(zbd_power,abs(2*pp[0]*frac_f0+pp[1])*np.sqrt(amp),'.-')
loglog(zbd_power,abs(2*pp[0]*frac_f0+pp[1])*np.sqrt(device-amp),'.-')

x = np.array([1e-7,5e-5])
loglog(x,np.sqrt(x)*3e-8)
grid()
```

```python
loglog(zbd_power,device,'o',label='100 Hz')
loglog(zbd_power,amp,'x',mew=2,label='7 kHz')
x = np.array([1e-6,5e-5])
loglog(x,x*3e-13,'--',lw=2,label='$\propto P$')
grid()
xlabel('Measured mm-wave power at source (W)')
ylabel('Fractional frequency noise (1/Hz)')
legend(loc='upper left',prop=dict(size='small'))
```

```python
loglog(zbd_power,device-amp,'o',label='data')
x = np.array([1e-6,5e-5])
loglog(x,x*8e-13,'--',lw=2,label='$\propto P$')
grid()
xlim(1e-8,1e-4)
ylim(1e-19,1e-16)
xlabel('Measured mm-wave power at source (W)')
ylabel('Corrected fractional frequency noise (1/Hz)')
legend(loc='upper left',prop=dict(size='small'))
```
