```python
def osemod(x,params):
    tau = params['tau'].value
    offs = params['offs'].value
    amp = params['amp'].value
    t0 = params['t0'].value
    #print tau,offs,amp,t0
    y = offs*np.ones_like(x)
    y[x>=t0] = amp*np.exp(-(x[x>=t0]-t0)/tau) + offs
    return y
```

```python
def oseobj(params,x,data):
    return osemod(x,params)-data
```

```python
def fitose(x,y):
    offs = y.min()
    amp = y.max()-y.min()
    print offs, amp
    params = lmfit.Parameters()
    params.add('tau',value=100,min=0,max=x.max())
    params.add('amp',value=amp,min=0)
    params.add('t0',value=0,min=x.min()/2,max=x.max()/2,vary=True)
    params.add('offs',value=offs, min = 0, max=amp*0.05,vary=True)
    result = lmfit.minimize(oseobj,params,args=(x,y),method='cobyla',
            options=dict(maxfev=1000000,maxiter=50000), tol = 1e-16)
#    mini = lmfit.Minimizer(oseobj,params,fcn_args=(x,y),xtol=1e-9,ftol=1e-9)
#    result = mini.scalar_minimize(method='Nelder-Mead')
    return result,params
```
