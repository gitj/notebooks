```python
import scipy.special
```

```python
Tc = 1.46
kBeV = 1.38e-23/1.6e-19
Delta = 1.74*kBeV*Tc
f0 = 100e6
h = 6.626e-34/1.6e-19
N0 = 1.72e10
```

```python
Tbase = 0.25
T = np.linspace(0.1,0.6,100)
P = np.linspace(1,10,1000)*1e-3
```

$$n_{qp,optical}(P_o) = \sqrt{\frac{N_{qp}\tau_{qp}\eta_o P_o}{\Delta}} = \sqrt{\frac{\kappa\eta_o P_o}{\Delta}}$$

Phil's spreadsheet estimates $\kappa=N_{qp}\tau_{qp}\sim$85

```python
nqp_opt = np.sqrt(85.*((0.5*P*1e-12)/(1.6e-19*Delta))) #85 is Nqp*tauqp
```

$$n_{qp,thermal}(T)=2N_0 \sqrt{2\pi k_B T \Delta}e^{-\Delta/(k_B T)}$$

```python
def nqp_therm(T):
    return 2*N0*np.sqrt(2*np.pi*kBeV*T*Delta)*np.exp(-Delta/(kBeV*T))
```

Gao et al. 2008 equation 16 gives $\sigma_1$ in terms of $n_{qp,total}$ and $T$

$$\frac{\sigma_1(n_{qp,total},T)}{\sigma_n} = \frac{2\Delta n_{qp} \sinh(\frac{hf}{2k_B T}) K_0(\frac{hf}{2k_B T})}{h f N_0 \sqrt{2\pi k_B T \Delta}}$$

So in the dark case, we can substitute $n_{qp,total}=n_{qp,thermal}$ to get

$$\frac{\sigma_{1,dark}(T)}{\sigma_n} = \frac{2\Delta 2N_0 \sqrt{2\pi k_B T \Delta}e^{-\Delta/(k_B T)} e^{-\Delta/(k_B T)} \sinh(\frac{hf}{2k_B T}) K_0(\frac{hf}{2k_B T})}{h f N_0 \sqrt{2\pi k_B T \Delta}e^{-\Delta/(k_B T)}} = \frac{4\Delta e^{-\Delta/(k_B T)} \sinh(\frac{hf}{2k_B T}) K_0(\frac{hf}{2k_B T})}{h f }$$

In the optical case, since $n_{qp,optical}\gg_{qp,thermal}$, we can substitute $n_{qp,total}=n_{qp,optical}$ to get

$$\frac{\sigma_{1,optical}(P_o,T)}{\sigma_n} = \frac{2\Delta \sqrt{\frac{\kappa\eta_o P_o}{\Delta}} \sinh(\frac{hf}{2k_B T}) K_0(\frac{hf}{2k_B T})}{h f N_0 \sqrt{2\pi k_B T \Delta}} = \frac{2 \sqrt{\kappa\eta_o P_o} \sinh(\frac{hf}{2k_B T}) K_0(\frac{hf}{2k_B T})}{h f N_0 \sqrt{2\pi k_B T}}$$

The same can be done for $\sigma_2$ (equation 17)

Then equation 20 gives the frequency and Q shifts:

$$\frac{\delta f }{f_0} \propto \frac{\delta \sigma_2}{\sigma_2} = \frac{\sigma_2(T,P)-\sigma_2(T_0,P_0)}{\sigma_2(T,P)}$$

$$\delta \frac{ 1 }{Q} \propto \frac{\delta \sigma_1}{\sigma_2} = \frac{\sigma_1(T,P)-\sigma_1(T_0,P_0)}{\sigma_2(T,P)}$$




```python
def sigma1(nqp,T):
    xi = (h*f0)/(2*kBeV*T)
    return ((2*Delta*nqp*np.sinh(xi)*scipy.special.k0(xi))/
            (h*f0*N0*np.sqrt(2*np.pi*kBeV*T*Delta)))
```

```python
def sigma2(nqp,T):
    xi = (h*f0)/(2*kBeV*T)
    return ((np.pi*Delta/(h*f0))*
            (1 - (nqp/(2*N0*Delta))*
             (1+np.sqrt((2*Delta)/(np.pi*kBeV*T))*np.exp(-xi)*scipy.special.i0(xi))))
```

```python
def sigma1_therm(T):
    return sigma1(nqp_therm(T),T)
```

```python
def sigma2_therm(T):
    return sigma2(nqp_therm(T),T)
```

```python
df_therm = (sigma2_therm(T)-sigma2_therm(T[0]))/sigma2_therm(T)
dq_therm = (sigma1_therm(T)-sigma1_therm(T[0]))/sigma2_therm(T)
df_opt = (sigma2(nqp_opt,0.2)-sigma2(nqp_opt[0],0.2))/sigma2(nqp_opt,0.2)
dq_opt = (sigma1(nqp_opt,0.2)-sigma1(nqp_opt[0],0.2))/sigma2(nqp_opt,0.2)
plot(1e6*df_therm,1e6*dq_therm,'.-',label='thermal')
plot(1e6*df_opt,1e6*dq_opt,'.-',label='optical')
xlim(-1000,0)
ylim(0,30)
print np.polyfit(df_therm,dq_therm,1)
print np.polyfit(df_opt,dq_opt,1)
xlabel('$\delta f / f_0$ [ppm]')
ylabel('$10^6/Q$')
legend(loc='upper right')
grid()
title("(Non?)-Equivalence of optical and thermal QPs")
```
