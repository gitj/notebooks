```python
for k in [1,8,15]:
    swp = swps[k]
    semilogy(swp.mmw_freq,swp.mmw_frac_response)
```

```python
avg = 0
n = 0
for k in [1,3,5,6,7,8,9,10,12,13,15]:
    swp = swps[k]
    semilogy(swp.mmw_freq,swp.mmw_frac_response,'gray')
    avg += swp.mmw_frac_response
    n+=1.
semilogy(swp.mmw_freq,avg/n,lw=2,color='r')
yscale('linear')
```

```python
plot(swp.mmw_freq,swp.zbd_voltage)
```

```python
avg = 0
n = 0
for k in [1,3,5,6,7,8,9,10,12,13,15]:
    swp = swps[k]
    semilogy(swp.mmw_freq/1e9,100*swp.mmw_frac_response/swp.zbd_voltage,'gray')
    avg += swp.mmw_frac_response
    n+=1.
semilogy(swp.mmw_freq/1e9,100*swps[9].mmw_frac_response/swp.zbd_voltage,lw=1,color='b')
semilogy(swp.mmw_freq/1e9,100*avg/n/swp.zbd_voltage,lw=2,color='r')
yscale('linear')
xlim(140,163)
ylim(0.0,25)
xlabel('Source frequency (GHz)')
ylabel('Detector response (arb)')
savefig('/home/gjones/paper_plots/2015-04-28-mmw-response-linear.pdf',bbox_inches='tight')
```
