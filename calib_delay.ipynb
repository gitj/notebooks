```python
import time
import kid_readout.utils.data_block
```

```python
fa = np.linspace(11.01,100,8)
fb = np.concatenate((fa,fa+0.001,fa+0.002,fa+0.003))
fb.sort()
fout = ri.set_tone_freqs(fb,nsamp=2**19)
ri._sync()
```

```python
dout = np.zeros(fout.shape,dtype='complex')
for k in range(4):
    ri.select_fft_bins(np.arange(len(ri.fft_bins))[k::4])
    time.sleep(0.2)
    d,addr = ri.get_data(2)
    dfilt = np.zeros((768,8))
    for m in range(8):
        dfilt[:,m] = kid_readout.utils.data_block.fftfilt.fftfilt(kid_readout.utils.data_block.lpf,d[:,m])[256:]
    dout[k::4] = d.mean(0)
```

```python
a1 = angle(dout).reshape((-1,4))
fout = fout.reshape((-1,4))
```

```python
plot(np.diff(unwrap(a1,axis=1),axis=1)/np.diff(fout,axis=1))
```
