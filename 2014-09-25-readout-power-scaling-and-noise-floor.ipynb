# Readout power scaling and LNA noise floor tests

Goal: Check that the noise floor scales as we expect as the readout power changes

### Setup

StarCryo 130919 0813f12 chip 

Power at output of analog conditioning for the 191 MHz tone measured to be -40 dBm when dac attenuator was in final state (11 dB)

File contains one sweep taken at 39 dB DAC atten. 16 simultaneous tones

Then timestreams were taken at each DAC attenuator level from 39 dB to 11 dB in 1 dB steps

### Calculations

Noise density at the input of the LNA should be around
$k_B T_N = -192\; \mathrm{dBm}/\mathrm{Hz}$
for LNA noise temperature of 4 K. So for the 191 MHz tone, if it is attenuated from -40 dBm by 40 dB of attenuators and 10 dB of cable loss, it would be at around -90 dBm, so the noise floor should be around -192 dBm/Hz - (-90) dBm = -102 dBc/Hz.

For 39 dB of DAC attenuation, the tone power would be about -118 dBm at the LNA, so the noise floor should be about -74 dBc/Hz

Looking at the measurement frequencies, index 9 is the one at 191 MHz that has no resonator. We'll also look at resonator index 0 later.

```python
rnc.timestream_20140925121818.measurement_freq[:16]
```

```python
figure(figsize=(12,8))
levs = []
for k,ts in enumerate(rnc.timestreams):
    data = ts.get_data_index(9)
    pxx,fr = mlab.psd(data,NFFT=2**10,Fs=ts.sample_rate[0])  # 2**10 NFFT gives nice spectra 
    # To normalize the spectra to dBc, we need to divide by the total carrier power
    pxx = pxx/(np.abs(data)**2).mean()
    plot(fr,10*np.log10(pxx),color=plt.cm.hsv(float(k)/(len(rnc.timestreams)+4)))
    levs.append(pxx[-100:].mean())  # grab the average of the highest 100 frequency points to represent the LNA noise floor
xlim(10,1e4)
xscale('log')
ylim(-110,-60)
grid()

annotate("11 dB DAC atten,\n-40 dBm",(100,-102),xytext=(-10,-30),textcoords='offset points',
         arrowprops=dict(width=1,frac=0.5),ha='right',va='top')
annotate("39 dB DAC atten,\n-68 dBm",(100,-75),xytext=(10,30),textcoords='offset points',
         arrowprops=dict(width=1,frac=0.5),ha='left')
xlabel('Frequency (Hz)')
ylabel('dBc/Hz')
title('Measured noise floor @ 191 MHz\nvarying tone power in 1 dB steps')
ax=gca()
ax.axhline(-103.0,linestyle='--',color='k')
text(20,-102,"-103 dBc/Hz",va='bottom')
```

```python
plot(attens,10*np.log10(levs),'.-')
xlabel('DAC Attenuator [dB]')
ylabel('LNA noise floor [dBc/Hz]')
grid()
```

```python
figure(figsize=(12,8))
levs2 = []
for k,ts in enumerate(rnc.timestreams):
    data = ts.get_data_index(0)
    pxx,fr = mlab.psd(data,NFFT=2**10,Fs=ts.sample_rate[0])
    pxx = pxx/(np.abs(data)**2).mean()
    plot(fr,10*np.log10(pxx),color=plt.cm.hsv(float(k)/(len(rnc.timestreams)+4)))
    levs2.append(pxx[-100:].mean())
xlim(10,1e4)
xscale('log')
ylim(-110,-60)

grid()
xlabel('Frequency (Hz)')
ylabel('dBc/Hz')
title('Measured noise spectra @ %.3f MHz\nvarying tone power in 1 dB steps' % ts.measurement_freq[0])
ax=gca()
ax.axhline(-103.0,linestyle='--',color='k')
text(20,-102,"-103 dBc/Hz",va='bottom')
```

```python
plot(attens,10*np.log10(levs2),'.-')
xlabel('DAC Attenuator [dB]')
ylabel('LNA noise floor [dBc/Hz]')
grid()
```
