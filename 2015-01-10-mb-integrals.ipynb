```python
x = np.logspace(-6,-2,1000)
```

```python
loglog(x,kid_readout.analysis.kid_response.fractional_freq_to_power(x, break_point=1e-4,scale=1))
```

```python
def fractional_freq_to_power(x, break_point, scale, exponent):
    return scale * ((x / break_point + 1) ** exponent - 1)
```

```python
for ex in [1,1.1,1.5,2,2.1,3,0.5]:
    loglog(x,fractional_freq_to_power(x, break_point=1e-4,scale=1,exponent=ex))
```
