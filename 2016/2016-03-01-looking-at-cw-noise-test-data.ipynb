```python
snm = snms[4]
```

```python
pca_evals = []
for snm in snms:
    pca_fr, S, evals, evects, angles, piq = kid_readout.analysis.timedomain.iqnoise.pca_noise(snm.fractional_fluctuation_timeseries,NFFT=2**16,Fs=snm.timeseries_sample_rate,
                                                                                 )
    pca_evals.append(evals)
```

```python
fig,axs = subplots(4,4,figsize=(18,18))
for k in range(16):
    ax = axs.flatten()[k]
    ax.loglog(pca_fr,pca_evals[k][1,:])
    ax.loglog(pca_fr,pca_evals[k][0,:])
    ax.set_ylim(1e-18,1e-16)
```
