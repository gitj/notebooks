$n_{qp}$: Quasiparticle density
==========================

The quasiparticle density, $n_{qp}$, in column F is calculated as:

$n_{qp} = 2 N_0 \sqrt{2\pi k_B T \Delta} \exp{\frac{-\Delta}{k_B T}} + n_{qp0}$

$N_0$ is the single spin density of states, 1.72e10 $\mu$m$^{-3}$eV$^{-1}$ for aluminum (cell B9)

$T$ is the physical temperature

$\Delta$ is taken as fixed, $1.74 k_B T_c$ (cell B11) with $T_c = 1.46$ K (cell B10)

$n_{qp0}$ is a constant, cell B24, and is asserted to be 5000

```python
T = np.linspace(0.1,0.5,500)
N0 = 1.72e10
kBeV = 1.38e-23/1.6e-19
Delta = 1.74 * kBeV * 1.46 #eV
nqp0 = 50
def nqp(T,nqp0=nqp0): 
    return 2 * N0 * np.sqrt(2*np.pi*kBeV*T*Delta) * np.exp(-Delta/(kBeV*T)) + nqp0
```

```python
for nqp0_ in [5,50,500,5000,50000]:
    semilogy(T,nqp(T,nqp0_),label=('%.1f' % nqp0_))
legend(loc = 'lower right',prop=dict(size='small'),title='$n_{qp0}$')    
xlabel('T [K]')
ylabel(r'''$n_{qp}$ [$\mu$m$^{-3}$]''')
title('Quasiparticle density')
grid()
```

$\tau_{qp}$: quasiparticle lifetime
=============================

$\tau_{qp}$ is simply scaled from a presumably empirical measurement at 350 mK, given in cell B18 as 40 $\mu$s. The scaling is:

$\tau_{qp} = 40\mu s \frac{n_{qp,350mK}}{n_{qp}(T)}$

```python
def tauqp(T):
    return 40e-6*nqp(.35)/nqp(T)
```

```python
plot(T,4*nqp(.35)/nqp(T))
ylabel(r'$\tau_{qp}$ [$\mu$s]')
xlabel('T [K]')
title('Quasiparticle lifetime')
grid()
```

Clearly the quasiparticle lifetime saturates at a level determined by $n_{qp0}$, in this case it happens to be about 80 $\mu$s

Effective potential $\mu$
=========================

Gao et al. 2008 Eqn 7 gives an expression for $n_{qp}$:

$n_{qp} = 2 N_0 \sqrt{2\pi k_B T \Delta} \exp{(-\frac{\Delta-\mu^{*}}{k_B T})}$

where $\mu^{*}$ is the effective chemical potential. This seems to be back-computed in column I from $n_{qp}$ calculated in column F, assuming $T$ in this equation is a fixed "Tbase" operating temperature given in cell B31 as 250 mK

```python
def mu(T,Tbase=0.25):
    return kBeV * T * np.log(nqp(T)/(2*N0*np.sqrt(2*np.pi*kBeV*T*Delta))) + Delta
```

```python
plot(T,mu(T))
ylabel('$\mu$ [eV]')
xlabel('T [K]')
title('Effective potential')
grid()
```

Complex admittivity
===================

The admittivity (a volume based property is):

$\sigma = \sigma_1 - j\sigma_2$

The conductivity is given in Gao 2008 Eqn 9 as:

$\sigma_1 = \sigma_n \frac{4\Delta}{\hbar\omega}\exp{(-\frac{\Delta-\mu^{*}}{k_B T})}\sinh{\xi}K_0(\xi)$

where

$\xi = \frac{\hbar \omega}{2 k_B T}$

And the susceptivity is given by Eqn 10 as:

$\sigma_2 = \sigma_n \frac{\pi \Delta}{\hbar \omega} (1-2\exp{(-\frac{\Delta-\mu^{*}}{k_B T} - \xi)} I_0(\xi))$

$\sigma_n$ is the normal conductivity, which comes from cell 6 and is fixed at 5e5 $\Omega^{-1}$cm$^{-1}$  **Note this is somewhat higher than the typical value listed for aluminum of 3.5e5**

$\sigma_1$ is caluclated in column K just as indicated, using $\mu$ from column I and $\xi$ from column H (based on the temperature)

$\sigma_2 = \frac{\sigma_n \pi \Delta}{\hbar \omega} (1-\frac{n_{qp}}{2 N_0 \Delta}(1+\sqrt{\frac{2\Delta}{\pi k_B T}}\exp{(-\xi(T))} I_0(\xi(T)))$

```python
f0 = 150e6
sigman = 1.6e5
Tbase = 0.25
def sigma1(T,Tbase = Tbase,sigman = sigman):
    xi = (6.626e-34*f0)/(2*1.38e-23*T)  #updated to use T instead of Tbase from Phil's corrections
    return ((sigman*(4*Delta*1.6e-19)/(6.626e-34*f0)) * 
            np.exp(-(Delta-mu(T))/(kBeV*T)) * 
            np.sinh(xi) * scipy.special.k0(xi)
            )

def sigma2(T,Tbase=Tbase,sigman=sigman):
    xi = (6.626e-34*f0)/(2*1.38e-23*T)
    return ((sigman*(np.pi*Delta*1.6e-19)/(6.626e-34*f0)) * 
            (1 - (nqp(T)/(2*N0*Delta))*(1+np.sqrt((2*Delta)/(np.pi*kBeV*T))*np.exp(-xi)*scipy.special.i0(xi)))
            )

```

$L_k$ the kinetic inductance
============================

$L_k$ is computed in column S using $\sigma_2$ from column L as:

$L_k = N_{sq}/(2\pi f_0 \sigma_2 h)$

Where $h$ is the film thickness, and $N_{sq}$ is the number of squares of the inductor of length $l$ and width $w$, computed in B13 as:

$N_{sq} = l/w$

The kinetic inductance fraction is

$\alpha = \frac{L_k}{L_k+L_g}$

```python
l = 55200 #um
w = 2 #um
h = 20 # nm
def Lk(T):
    nsq = l/float(w)
    return nsq/(2*np.pi*f0*sigma2(T)*h * 1e-7)  #1e-7 is converting nm to cm
```

```python
plot(T,Lk(T)*1e9)
ylabel('Lk [nH]')
xlabel('T [K]')
title('kinetic inductance')
grid()
```

```python
Lg = 7e-8 # Henry
plot(T,Lk(T)/(Lk(T)+Lg))
title('kinetic inductance fraction')
ylabel(r'$\alpha$')
xlabel('T [K]')
grid()
```

Qi
===

Qi is calculated in column M as:

$Q_i = (2\pi f_0(L_g + L_k)) / (N_{sq}\sigma_1/(\sigma_2^2 h))$

Phil explains this formula as follows:

$Q_i = 2\pi f_0 L_{total} / R_{total}$

The resistivity is:

$\rho = 1/(\sigma_1 + j\sigma_2) = (\sigma_1 - j\sigma_2)/(\sigma_1^2+\sigma_2^2)$

Assuming $\sigma_2 \gg \sigma_1$:

$\rho = \frac{\sigma_1}{\sigma_2^2} - j\frac{1}{\sigma_2}$

So the $(N_{sq}/h)(\sigma_1/\sigma_2^2)$ portion is $R_{total}$

```python
Lg = 1.26e-7 # Henry
def Qi(T,Lg=Lg):
    nsq = l/w
    react = 2*np.pi*f0*(Lg+Lk(T))
    diss = nsq*sigma1(T)/(sigma2(T)**2*h*1e-7)
    return react/diss
```

```python
semilogy(T,Qi(T))
title('internal Q')
ylabel('$Q_i$')
xlabel('T [K]')
grid()
```

======

$Q_r$ is calculated in column N using a fixed $Q_c$ given in B20 as 1e5

```python
Qc = 1.5e5
def Qr(T,Qc=Qc):
    return 1/(1/Qi(T)+1/Qc)
```

Resonance shift and number of quasiparticles
============================================

The total number of quasiparticles $N_{qp}$ is calculated in column Q by multiplying $n_{qp}$ by the inductor volume

The resonance frequency is as usual $1/(2\pi\sqrt{(L_k+L_g)C})$

The resonance frequency as a function of temperature is used with $N_{qp}(T)$ to numerically compute $\frac{df}{dN_{qp}}$

```python
V = w*l*h*1e-3 # um^3, 1e-3 to convert h from nm to um
C = 11.7e-12  # Farads
def Nqp(T,V=V):
    return nqp(T)*V
def fres(T,C=C,Lg=Lg):
    return 1/(2*np.pi*np.sqrt((Lk(T)+Lg)*C))

def dfdNqp(T):
    x0 = fres(T)
    x1 = fres(T+1e-3)
    y0 = Nqp(T)
    y1 = Nqp(T+1e-3)
    return -(x1-x0)/(y1-y0)
```

```python
plot(T,(fres(T)/fres(0.001) - 1)*1e6)
ylabel('Fractional frequency shift [ppm]')
xlabel('T [K]')
grid()
```

```python
plot(T,dfdNqp(T))
ylim(0.00,0.005)
xlabel('T [K]')
ylabel('d$f$/d$N_{qp}$ [Hz/count]')
grid()
```

G-R noise
=========

The G-R noise level in Hz/$\sqrt{}$Hz is calculated in column V, labeled ef(G-R), as

$GR = 2\sqrt{N_{qp}\tau_{qp}}\frac{df}{N_{qp}}$

Dividing by $f_0$ gives the fractional noise in 1/$\sqrt{}$Hz in column W

Squaring the fractional noise gives an NEP like value in column X

```python
def Ngr(T):
    return 2*np.sqrt(Nqp(T)*tauqp(T))*dfdNqp(T)
```

```python
plot(T,Ngr(T)**2)
ylim(0.0,.08)
title('G-R noise level')
ylabel('Hz$^2$/Hz')
xlabel('T [K]')
grid()
```

Voltage responsivity
====================

To express the amplifier noise in terms of a frequency shift, we need dV/df, which Phil computes from dS21 in column Y as:

$\frac{dV}{df} = 4 V_{in} \alpha Q_r^2/(Q_c f_0(T))$

$\alpha = \frac{L_k}{L_k+L_g}$

This is derived as:

$S_{21} = 1-\frac{Q_r}{Q_c}\frac{1}{1+2jQ_r(f-f_0)/f_0}$

$V_{out} = 2\alpha V_{in}S_{21}$

$\frac{dS_{21}}{df} = \frac{2jQ_r^2}{Q_c f_0 (2jQ_r(f-f_0)/f_0 + 1)^2}$

Near resonance, this becomes:

$\frac{dS_{21}}{df} = \frac{2jQ_r^2}{Q_c f_0}$

The readout voltage $V_{in}$ is derived from the power in dBm as expected assuming 50 ohms:

$V = \sqrt{50 \cdot 10^{PdBm/10} \cdot 10^{-3}}$

```python
Pread = -76 # dBm
Vread = np.sqrt(50*10**(Pread/10)*1e-3)
def dVdf(T,Vread=Vread):
    alpha = Lk(T)/(Lk(T)+Lg)
    return 4*Vread*alpha*Qr(T)**2/(Qc*fres(T))
```

```python
plot(T,dVdf(T))
title('Voltage responsivity')
ylabel('dV/df [V/Hz]')
xlabel('T [K]')
grid()
```

Amplifier noise
===============

We can express the amplifier noise temperature in kelvin as a power using the Raleigh Jeans approximation:

$P_{amp} = k_B T_{noise}$

And the RMS voltage is then:

$V_{noise} = \sqrt{4 k_B T_{noise} 50}$

Then dividing by dV/df we get the amplifier noise in Hz^2/Hz in column Z

```python
Tn = 4.0
Vn = np.sqrt(4*1.38e-23*Tn*50)
print Vn
def Namp(T):
    return Vn/dVdf(T)
```

```python
semilogy(T,Namp(T)**2,label='Amplifier')
semilogy(T,Ngr(T)**2,label = 'G-R')
ylabel('Hz$^2$/Hz')
xlabel('T [K]')
legend(loc='upper left')
grid()
title('noise contributions')
ylim(0,0.5)
```

```python
def noise(f,T):
    fqp = 1/tauqp(T)
    fr = 150e6/Qr(T)
    gr = (Ngr(T)*abs(1/(1+1j*f/fr)))**2
    return gr + Namp(T)**2
```

```python
ff = np.logspace(2,6,1000)
for T_ in np.linspace(0.2,0.35,5):
    loglog(ff,noise(ff,T_),label=('T=%.2f K' % T_))
xlabel('frequency [Hz]')  
title('simulated noise spectra')
ylabel('Hz$^2$/Hz')
legend(loc='lower left')
grid()
```

TLS effect on frequency shift
=============================
From Jonas' review article, equation 25 is:

$\frac{\delta \omega_r}{\omega_r} = \frac{F_{TLS} \delta_0}{\pi}[\Re(\Psi(0.5 + \frac{\xi}{2j\pi})) - \ln(\xi)]$

```python
def TLS(T,F_TLS=1,delta0=1):
    xi = (6.626e-34*f0)/(2*1.38e-23*T)
    return ((F_TLS*delta0/np.pi) *
            (np.real(scipy.special.psi(0.5+(xi/(2j*np.pi))))
             -np.log(xi)))
```

```python
F_TLS = 1e-5
plot(T,TLS(T,F_TLS=F_TLS)*1e6,'--',label='TLS contribution')
fnew = fres(T)*(1+TLS(T,F_TLS=F_TLS))
plot(T,(fres(T)/fres(1e-3)-1)*1e6,'--',label='No TLS')
plot(T,(fnew/fres(1e-3)-1)*1e6,label='Total shift')
ylabel('fractional frequency shift [ppm]')
xlabel('T [K]')
legend(loc='lower left')
grid()
title('TLS effect on frequency shift')
ylim(-400,100)
```

TLS effect on $Q_i$
===================

Loss tangent is given by Jonas' equation 24:

$\delta_{TLS} = \delta_0 \tanh(\xi)$

```python
def deltaTLS(T,delta0=1e-5):
    xi = (6.626e-34*f0)/(2*1.38e-23*T)
    return delta0*np.tanh(xi)
    
```

```python
plot(T,1/(deltaTLS(T,1e-4)+1/Qi(T)))
plot(T,1/(deltaTLS(T,1e-4)))
plot(T,Qi(T))
yscale('log')
grid()
```
