```python
def normalize_f0(x):
    x['f_0_max'] = x['noise_off_f_0'].max()
    x['noise_on_frac_f0'] = (x['f_0_max'] - x['noise_on_f_0']) / x['f_0_max']
    x['noise_off_frac_f0'] = (x['f_0_max'] - x['noise_off_f_0']) / x['f_0_max']
    return x
```

```python
def fit_response(x):
    x['zbd_power'] = x['zbd_voltage']/2200.
    mask = (x.sweep_primary_load_temperature==0) & (x.zbd_power > 0)
    pp = np.polyfit(np.log10(x[mask]['noise_on_frac_f0']),np.log10(x[mask]['zbd_power']),1)
    x['pp0']=pp[0]
    x['pp1']=pp[1]
    #x['watts_to_ppm'] = (2*pp[0]*x['noise_on_frac_f0']+pp[1])
    x['noise_on_reconstructed_power'] = 10**np.polyval(pp,np.log10(x['noise_on_frac_f0']))
    x['noise_off_reconstructed_power'] = 10**np.polyval(pp,np.log10(x['noise_off_frac_f0']))
    return x
                                            
```

```python
dfr2 = dfr.groupby(['resonator_id']).apply(fit_response)
```

```python
import kid_readout.analysis.mcfit
reload(kid_readout.analysis.mcfit)
import kid_readout.analysis.kid_response
reload(kid_readout.analysis.kid_response)
def fit_response_mcmc(x):
    x['zbd_power'] = x['zbd_voltage']/2200.
    mask = (x.sweep_primary_load_temperature==0) & (x.zbd_power > 0)
    fit = kid_readout.analysis.kid_response.MCMCKidResponseFitter(x[mask]['noise_on_frac_f0'],x[mask]['zbd_power'],
                                                                  errors = np.where(x[mask]['zbd_power']>1e-7,
                                                                                    x[mask]['zbd_power']*1e-2,
                                                                                    1e-8))
    fit.run(burn_in=400)
    #blah = fit.triangle()
    x['response_break_point']=fit.mcmc_params['break_point'].value
    x['response_scale']=fit.mcmc_params['scale'].value
    x['response_break_point_err']=fit.mcmc_params['break_point'].stderr
    x['response_scale_err']=fit.mcmc_params['scale'].stderr
    #x['watts_to_ppm'] = (2*pp[0]*x['noise_on_frac_f0']+pp[1])
    x['noise_on_reconstructed_power'] = fit.to_power(x['noise_on_frac_f0'])
    x['noise_off_reconstructed_power'] = fit.to_power(x['noise_off_frac_f0'])
    return x
```

```python
import scipy.integrate
def Bnu(nu,T,emissivity=1.0):
    return emissivity*6.626e-34*nu/(np.exp((6.626e-34/1.38e-23)*(nu/T))-1)
def P(T,low=120e9,high=170e9):
    def _P(T):
        return scipy.integrate.quad(Bnu,low,high,args=(T))[0]
    return np.vectorize(_P)(T)
```
