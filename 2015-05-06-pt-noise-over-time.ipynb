```python
df = pd.DataFrame(columns=['epoch',
                           'resonator_index',
                           'chip_id',
                           'optical_state',
                           'description',
                           'dac_atten',
                           'power_dbm',
                           'sxx_harm1',
                           'syy_harm1',
                           'sxx_harm2',
                           'syy_harm2',
                           'sxx_110',
                           'syy_110',
                           'sxx_med110',
                           'syy_med110',
                           'is_dark',
                           'pkl_file',
                           'Q',
                           'Qi',
                           'a'
                           ])
idx = 0
for fn in pkls:
    try:
        pkl =kid_readout.analysis.archive.load_noise_pkl(fn)
    except:
        print "couldn't unpickle",fn
        continue
        nm = kid_readout.analysis.noise_measurement.SweepNoiseMeasurement('/home/data/'+os.path.splitext(os.path.split(fn)[1])[0]+'.nc')
    if type(pkl) is dict:
        if len(pkl['noise_on_measurements']):
            nms = pkl['noise_on_measurements']
        else:
            print fn,"has no noise on measurements"
            continue
    else:
        nms = pkl
    if not len(nms):
        print fn,"has no nms"
        continue
    
    for nm in nms:
        if hasattr(nm,'pca_freq'):
            pca_freq = nm.pca_freq
            pca_eigvals = nm.pca_eigvals
        elif hasattr(nm,'pca_fr'):
            pca_freq = nm.pca_fr
            pca_eigvals = nm.pca_evals
        else:
#            print fn,"has no pca"
            continue
        idx_1_harm = np.argmin(np.abs(pca_freq-1.4))
        idx_2_harm = np.argmin(np.abs(pca_freq-2.8))
        idx_110 = np.argmin(np.abs(pca_freq-110))
        try:
            epoch = nm.timestream_epoch
        except:
            try:
                epoch = nm.ts_epoch
            except:
                print "no epoch for",fn
                print dir(nm)
                7/0
        try:
            res_id = nm.resonator_index
        except:
            res_id = nm.index
        try:
            params = nm.fit_params
        except:
            params = nm.params
        if 'a' in params:
            a = params['a'].value
        else:
            a = 0
    
        info = kid_readout.analysis.resources.experiments.get_experiment_info_at(epoch)
        df.loc[idx] = {'epoch' : epoch,
                       'resonator_index' : res_id,
                       'chip_id' : info['chip_id'],
                       'optical_state' : info['optical_state'],
                       'description' : info['description'],
                       'dac_atten' : nm.atten,
                       'power_dbm' : nm.power_dbm,
                       'sxx_harm1' : pca_eigvals[1,idx_1_harm],
                       'syy_harm1' : pca_eigvals[0,idx_1_harm],
                       'sxx_harm2' : pca_eigvals[1,idx_2_harm],
                       'syy_harm2' : pca_eigvals[0,idx_2_harm],
                       'sxx_110' : pca_eigvals[1,idx_110],
                       'syy_110' : pca_eigvals[0,idx_110],
                       'sxx_med110' : np.median(pca_eigvals[1,:idx_110]),
                       'syy_med110' : np.median(pca_eigvals[0,:idx_110]),
                       'is_dark' : (info['optical_state']=='dark'),
                       'pkl_file' : fn,
                       'Q' : params['Q'].value,
                           'Qi' : nm.Q_i,
                           'a': a,}
        idx += 1
    del nm
    del pkl
```
