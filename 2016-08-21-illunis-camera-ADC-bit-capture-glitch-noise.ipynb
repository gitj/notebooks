```python
import scipy.ndimage
```

```python
im = np.load('/home/pmcroot/data1/2016-08-20_50mm_star_focus_test_650ms/2016-08-20_234111_focus_step_0009.npz')['image']/16
```

```python
_=hist(im.flatten(),bins=100,log=True)
xlabel('Pixel value (12-bit ADU)')
ylabel('counts')
suptitle('Log histogram of pixel values')
```

```python
figure(figsize=(18,12))
imshow(im[::4,::4],interpolation='nearest',cmap=cm.gray,extent=[0,im.shape[1],im.shape[0],0])
clim(0,80)
colorbar()
title('Downsampled image showing inhomogenous noise\nFree running exposure set to 65000 = 650 ms')
```

```python
figure(figsize=(18,12))
imshow(im[1000:1500,1600:2200],interpolation='nearest',cmap=cm.gray,extent=[1600,2200,1500,1000])
clim(0,80)
colorbar()
axhline(1250)
title('Zoom on one region. Blue line shows slice used in following plots')
```

```python
figure(figsize=(18,6))
plot(im[1250,:])
title('Slice along row 1250 of image showing several glitching regions')
```

```python
figure(figsize=(18,6))
plot(im[1250,:])
xlim(1600,2200)
axvspan(1600,1700,alpha=0.1)
axvspan(1960,2015,alpha=0.1)
title('Zoom of slice along row 1250 with details of glitching regions highlighted')
```

```python
figure(figsize=(18,12))
imshow(scipy.ndimage.gaussian_filter(im[1000:1500,1600:2200],2),interpolation='nearest',cmap=cm.gray,extent=[1600,2200,1500,1000])
clim(0,80)
colorbar()
title('Zoom on same region after applying 2 pixel gaussian filter')
```
