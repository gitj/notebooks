```python
dark_beta = []
dark_beta_err = []
t200mk_beta = []
light_beta = []
f0s = []
for k in range(20):
    r0 = dfr[(dfr.atten==39.)&(dfr.resonator_index==k)&(dfr.timestream_duration>6.)]
    if len(r0) < 1:
        continue
    dark = r0[r0.optical_load=='dark']
    t200mk = r0[(abs(r0.sweep_primary_package_temperature-0.2)<0.01)
                & (r0.optical_load != 'dark')]
    light = r0[(r0.optical_load!='dark')&(abs(r0.sweep_primary_load_temperature-4.0)<0.1)]

    fit,cov = np.polyfit(dark.fractional_delta_f_0*1e6,1e6/dark.Q_i,1,cov=True)
    dark_beta.append(0.5/fit[0])
    dark_beta_err.append(np.sqrt(1/cov[1,1])*fit[0])
    fit = np.polyfit(t200mk.fractional_delta_f_0*1e6,1e6/t200mk.Q_i,1)
    t200mk_beta.append(0.5/fit[0])
    fit = np.polyfit(light.fractional_delta_f_0*1e6,1e6/light.Q_i,1)
    light_beta.append(0.5/fit[0])
    f0s.append(r0.f_0_max.iat[0])
f0s = np.array(f0s)
light_beta = np.abs(np.array(light_beta))
dark_beta = np.abs(np.array(dark_beta))
t200mk_beta = np.abs(np.array(t200mk_beta))
dark_beta_err = np.array(dark_beta_err)
order = f0s.argsort()
f0s = f0s[order]
light_beta = light_beta[order]
dark_beta = dark_beta[order]
dark_beta_err = dark_beta_err[order]
t200mk_beta = t200mk_beta[order]
msk = ((light_beta>24)|(dark_beta>24))&(f0s>100)
f0s = f0s[~msk]
dark_beta = dark_beta[~msk]
dark_beta_err = dark_beta_err[~msk]
t200mk_beta = t200mk_beta[~msk]
light_beta = light_beta[~msk]

figure(figsize=(5.5,4))
plot(f0s,light_beta,'o-',lw=1,label='$T_{bb}$ = 4 K, $T_{base}$ sweep')
errorbar(f0s,dark_beta,marker='v',yerr=dark_beta_err,lw=1,label='Dark, $T_{base}$ sweep')
plot(f0s,t200mk_beta,'+-',mew=2,lw=1,markersize=8,label='$T_{base}$ = 200 mK, $T_{bb}$ sweep')

ff = np.linspace(75,165)
s1,s2 = kid_readout.analysis.kid_eqns.s1s2(0.2,1.46,ff*1e6)
plot(ff,0.5*s2/s1,lw=2,label='Theory, $T_{base}$ = 200 mK')
xlabel('Resonant Frequency [MHz]')
ylabel(r'$\beta$',size='x-large')
xlim(75,165)
ylim(5,35)
grid()
legend()

savefig('/home/gjones/paper_plots/sc5x4_betas.pdf',bbox_inches='tight')
```
