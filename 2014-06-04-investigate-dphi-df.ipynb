```python
from kid_readout.analysis import resonator,khalil
```

```python
def myguess(x,y):
    p = khalil.bifurcation_guess(x,y)
    p['a'].max=4.0
    return p
```

```python
rr = resonator.Resonator(nm.sweep_freqs_MHz,nm.sweep_s21,errors=nm.sweep_errors,guess=myguess,model=khalil.bifurcation_s21)
```

```python
rr.a
```

```python
plot(nmod.real,nmod.imag)
plot(nm.sweep_normalized_s21.real[rr.mask],nm.sweep_normalized_s21.imag[rr.mask],'.-')
plot(nmod[abs(nm.sweep_model_freqs_MHz-rr.f_0).argmin()].real,nmod[abs(nm.sweep_model_freqs_MHz-rr.f_0).argmin()].imag,'o')
```
