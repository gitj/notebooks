```python
from kid_readout.utils.easync import EasyNetCDF4
```

```python
nc = EasyNetCDF4('/home/data/2013-12-09_113911.nc')
```

```python
tsa = nc.timestreams.timestream_20131209114251.data[:].view('complex').squeeze()
```

```python
tsb = nc.timestreams.timestream_20131209114422.data[:].view('complex').squeeze()
```

```python
tsa.shape
```

```python
figure(figsize=(12,6))
paa,fr = mlab.psd(tsa,NFFT=2**18,Fs=256e6/2**11)
pbb,fr = mlab.psd(tsb,NFFT=2**18,Fs=256e6/2**11)
plot(fr,10*np.log10(paa),label='PT On')
plot(fr,10*np.log10(pbb),label='PT Off')
xscale('symlog')
xlabel('Frequency (Hz)')
ylabel('Power dB/Hz')
legend(loc='upper right')
title('PT effect 2013-12-09')
grid()
```

```python
figure(figsize=(12,6))
paa,fr = mlab.psd(tsa,NFFT=2**22,Fs=256e6/2**11)
pbb,fr = mlab.psd(tsb,NFFT=2**22,Fs=256e6/2**11)
plot(fr,10*np.log10(paa),label='PT On')
plot(fr,10*np.log10(pbb),label='PT Off')
xscale('symlog')
xlim(0.01,200)
xlabel('Frequency (Hz)')
ylabel('Power dB/Hz')
legend(loc='upper right')
pks = np.flatnonzero(10*np.log10(paa) > 12)
ax = gca()
for pk in pks:
    ax.text(fr[pk],10*np.log10(paa[pk]),('%.1f' % (fr[pk]/1.4)),size='xx-small')
title('PT effect 2013-12-09. First few harmonics of 1.4 Hz labled',size='small')
grid()
```
