```python
import scipy.integrate
```

First, define the blackbody curve:

$$ \frac{2h\nu^3}{c^2 e^{h\nu/kT}-1} \frac{n_{modes}}{\lambda^2} $$

Since we have the low pass filter in place, we'll assume $n_{modes} = 1$ giving:

$$ B_{\nu}(T) = \frac{2h\nu}{e^{h\nu/kT}-1} $$


```python
def Bnu(nu,T):
    return 2*6.626e-34*nu/(np.exp((6.626e-34/1.38e-23)*(nu/T))-1)
```

Now integrate over the band, from 120 GHz to 170 GHz, to get power as a function of temperature

```python
def P(T):
    def _P(T):
        return scipy.integrate.quad(Bnu,120e9,170e9,args=(T))[0]
    return np.vectorize(_P)(T)
```

```python
T = np.linspace(1,10,200)
```

```python
semilogy(T,P(T))
title("Blackbody power vs. temperature\nfor a single mode from 120 to 170 GHz")
ylabel('Power (W)')
xlabel('Load Temperature (K)')
grid()
```

Now define the photon NEP using:

$$ NEP_{photon}(T) = \sqrt{2h\nu P_o(T) (1+\frac{P_o(T)}{2h\nu\Delta\nu}}) $$

$P_o$ will be multiplied by "frac" to represent the fraction of power actually absorbed by the detector.

The term $\frac{P_o(T)}{2h\nu\Delta\nu}$ is the photon occupation number and can be thought of as the number of photons $P_o/(h\nu)$ present during one coherence time $1/(2\Delta\nu)$

```python
def NEP(T,frac=0.25):
    return np.sqrt(2*6.626e-34*150e9*P(T)*frac*(1+P(T)*frac/(2*6.626e-34*150e9*40e9)))
```

```python
for frac in np.linspace(0.01,0.9,5):
    plot(T,NEP(T,frac=frac),color=plt.cm.spectral(frac),label=('%.0f%%' % (frac*100)))
title('Photon NEP versus load temperature\nand absorbed power fraction')
ylabel('NEP ($W/\sqrt{Hz}$)')
xlabel('Load Temperature (K)')
legend(loc='upper left',ncol=5,prop=dict(size='x-small'))
grid()
```

To convert to NET, we calculate $dP/dT$ to convert between watts and kelvin. We do this with a simple numerical derivative

```python
def dPdT(T):
    return (P(T+1e-3)-P(T))/1e-3
```

```python
plot(T,dPdT(T))
title('$dP/dT$ versus load temperature')
ylabel('$dP/dT$ (W/K)')
xlabel('Load Temperature (K)')
grid()
```

Now NET is just

$$NET = NEP\frac{dP}{dT}^{-1}$$

```python
def NET(T,frac=0.5):
    return NEP(T,frac=frac)/dPdT(T)
```

```python
for frac in np.linspace(0.01,0.9,5):
    plot(T,1e6*NET(T,frac=frac)/np.sqrt(2),
         color=plt.cm.spectral(frac),label=('%.0f%%' % (frac*100)))
title('$NET_{photon}$ vs. absorption fraction')
ylabel('Photon NET ($\mu K \sqrt{s}$)')
xlabel('Load Temperature (K)')
legend(loc='upper left',ncol=5,prop=dict(size='x-small'))
ylim(0,50)
grid()
```

Now define total NEP with contributions from recombination and a possible constant "device noise". Thre recombination noise is a scaled version of the photon NEP (scaling is 1 by default)

```python
def NEP_total(T,recomb_factor=1,extra=0,frac=0.5):
    NEP_phot = NEP(T,frac=frac)
    NEP_device = NEP_phot*recomb_factor
    return np.sqrt(NEP_phot**2 + NEP_device**2 + extra**2)
```

```python
for frac in np.linspace(0.01,0.9,5):
    plot(T,1e6*NEP_total(T,frac=frac,extra=0)/(dPdT(T)*np.sqrt(2)),
         color=plt.cm.spectral(frac),label=('%.0f%%' % (frac*100)))
title('$NET_{total}$ vs. absorption fraction')
ylabel('Total NET ($\mu K \sqrt{s}$)')
xlabel('Load Temperature (K)')
ylim(0,50)
legend(loc='upper left',ncol=5,prop=dict(size='x-small'))
grid()
```

```python
for frac in np.linspace(0.01,0.9,5):
    plot(T,1e6*NEP_total(T,frac=frac,extra=2e-17)/(dPdT(T)*np.sqrt(2)),
         color=plt.cm.spectral(frac),label=('%.0f%%' % (frac*100)))
title('$NET_{total}$ vs. absorption fraction\nwith additional 2e-17 device noise')
ylabel('Total NET ($\mu K \sqrt{s}$)')
xlabel('Load Temperature (K)')
ylim(0,50)
legend(loc='upper left',ncol=5,prop=dict(size='x-small'))
grid()
```

```python
figure(figsize=(9,6))
for frac in np.linspace(0.01,0.6,5):
    plot(T,1e6*NEP_total(T,frac=0.1,extra=frac*1e-16)/(dPdT(T)*np.sqrt(2)),
         color=plt.cm.spectral(frac/0.7), label=('%.2g' % (frac*1e-16)))
title('$NET_{total}$ vs. device noise. 10% absorbtion')
ylabel('Total NET ($\mu K \sqrt{s}$)')
xlabel('Load Temperature (K)')
ylim(0,50)
grid()
legend(loc='lower left',ncol=5,prop=dict(size='x-small'))
```
