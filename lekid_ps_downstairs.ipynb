```python
d = nc.timestreams.timestream_20131023182155.data[:].view('complex128')
d.shape
```

```python
dmag = np.abs(d)
dph = np.unwrap(np.angle(d))
```

```python
fsadc = nc.timestreams.timestream_20131023182155.fs[:]
tones = nc.timestreams.timestream_20131023182155.tone[:]
nsamps = nc.timestreams.timestream_20131023182155.nsamp[:]
resf0s = (fsadc*tones)/nsamps
```

```python
f = figure(figsize=(12,8))
for k in range(4):
    ax = f.add_subplot(2,2,k+1)
    ax.hist(np.abs(d[k,:]),bins=100)
    ax.set_title('%.3f MHz' % resf0s[k])
```

```python
psds = []
psdsdt = []
for det in range(4):
    pxx,fr = mlab.psd(d[det,:],NFFT=2**18,Fs=256e6/2**14)
    psds.append(pxx[None,:])
    pxx,fr = mlab.psd(d[det,:]-d[det,:].mean(),NFFT=2**18,Fs=256e6/2**14)
    psdsdt.append(pxx[None,:])
psds = np.concatenate(psds,axis=0)
psdsdt = np.concatenate(psdsdt,axis=0)
```

```python
posmsk = fr>0
negmsk = fr<0
```

```python
f = figure(figsize=(12,8))
for k in range(4):
    ax = f.add_subplot(2,2,k+1)
    ax.loglog(fr[posmsk],psdsdt[k,:][posmsk])
    ax.loglog(-fr[negmsk],psdsdt[k,:][negmsk])
    spurs = [1.2,2.4,3.6,60,120]
    for spf in spurs:
        ax.plot([spf,spf],[1e-2,1e-1],'k')
    ax.set_ylabel('PSD')
    ax.set_ylim(1e-4,1e0)
    ax.grid()
    ax.set_title('%.3f MHz' % resf0s[k])
ax.set_xlabel('Hz')
```

```python
dtd = d - d.mean(1)[:,None]
```

```python
xpsd = []
xlabs = []
for k in range(3):
    for m in range(k+1,4):
        xlabs.append((k,m))
        pxy,fr = mlab.csd(dtd[k,:],dtd[m,:],NFFT=2**18,Fs=256e6/2**14)
        xpsd.append(pxy)
```

```python
xlabs
```

```python
f = figure(figsize=(12,8))
for k in range(6):
    ax = f.add_subplot(2,3,k+1)
    ax.loglog(fr[posmsk],np.abs(xpsd[k][posmsk]))
    ax.loglog(-fr[negmsk],np.abs(xpsd[k][negmsk]))
    spurs = [1.2,2.4,3.6,60,120]
    for spf in spurs:
        ax.plot([spf,spf],[1e-2,1e-1],'k')
    
    ax.set_ylim(1e-5,1e0)
    ax.set_ylabel('xPSD')
    ax.grid()
    ax.set_title(str(xlabs[k]))
ax.set_xlabel('Hz')
```
