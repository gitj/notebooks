```python
def s21(f,f0,Qc,Q,epsilon=0,beta=1.0):
    return 1-(1/Qc)/(1/Q + beta*epsilon + 2j*(f-f0*(1-epsilon))/(f0*(1-epsilon)))
```

```python
f = np.linspace(99.9,100.1,1000)
f0 = 100.0
Qc = 1e4
Qi = 1e5
Q = np.real(1/(1/Qc + 1/Qi))
```

```python
y=s21(f,f0,Qc,Q)
plot(y.real,y.imag,'.-')
```

```python
def plot_s21(f0,Qc,Q,epsilon=0,beta=1.0,fstar=None):
    if fstar is None:
        fstar = f0
    y = s21(f,f0,Qc,Q,epsilon=epsilon,beta=beta)
    ystar = s21(fstar,f0,Qc,Q,epsilon=epsilon,beta=beta)
    plot(y.real,y.imag)
    plot(ystar.real,ystar.imag,'x',mew=2)
```

```python
for eps in np.linspace(0,1e-4,10):
    plot_s21(f0,Qc,Q,epsilon=eps,beta=1)
```

```python
y=s21(f0,f0,Qc,Q,epsilon=np.linspace(-1e-4,1e-4,100),beta=.1)
ybase = s21(f,f0,Qc,Q)
```

```python
plot(y.real,y.imag,'.-')
plot(ybase.real,ybase.imag)
```
