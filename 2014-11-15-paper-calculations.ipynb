```python
def Bnu(nu,T,emissivity=0.92):
    return 2*emissivity*6.626e-34*nu/(np.exp((6.626e-34/1.38e-23)*(nu/T))-1)
```

```python
def P(T,low=130e9,high=180e9):
    def _P(T):
        return scipy.integrate.quad(Bnu,low,high,args=(T))[0]
    return np.vectorize(_P)(T)
```

```python
def plot_deltaf(nstar,tau_max_us,eta):
    Tref=4.0
    T = np.linspace(1,12,1000)
    plot(T,-1e6*deltaf_T(T,nstar=nstar,taumax=tau_max_us*1e-6,eta=eta),label='theory')
    #plot(T,-1e6*xPo(P(T),nstar=nstar,taumax=tau_max_us*1e-6,eta=eta),label='theory')
#    plot(T,-1e6*deltaf_T(T,nstar=nstar,tau_max=tau_max_us*1e-6,eta=eta,alpha=0.35),label='theory')
    plot(T,-40*T+40*Tref,lw=4,alpha=.5,label='observed trend')
    ylim(-200,100)
    xlabel("Load Temperature [K]")
    ylabel("Fractional frequency shift relative to %.1f K [ppm]"%Tref)
    legend(loc='lower left')
    grid()
    title("nstar=%.1f, tau_max=%.1f us, eta=%.1f" % (nstar,tau_max_us,eta))
```
