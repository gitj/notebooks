```python
import scipy.integrate
def Bnu(nu,T,emissivity=1.0):
    return emissivity*6.626e-34*nu/(np.exp((6.626e-34/1.38e-23)*(nu/T))-1)
def P(T,low=120e9,high=170e9):
    def _P(T):
        return scipy.integrate.quad(Bnu,low,high,args=(T))[0]
    return np.vectorize(_P)(T)
```

```python
def Bnu2(nu,T,emissivity=1.0):
    return ((nu/1e9)**2/1.95e4)*emissivity*6.626e-34*nu/(np.exp((6.626e-34/1.38e-23)*(nu/T))-1)
def P2(T,low=120e9,high=170e9):
    def _P(T):
        return scipy.integrate.quad(Bnu2,low,high,args=(T))[0]
    return np.vectorize(_P)(T)
```
