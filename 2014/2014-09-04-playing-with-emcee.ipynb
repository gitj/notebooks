```python
def sinusoid(params,x):
    amp = params['amp'].value
    period = params['period'].value
    phase = params['phase'].value
    return amp*np.sin(2*np.pi*x/period + phase)
```

```python
def light_curve_model(params,x):
    return sinusoid(params, x) + params['mean'].value
```

```python
ndata = 32
outlier_fraction = 0.1
true_params = lmfit.Parameters()
true_params.add('amp',value=1.5)
true_params.add('period',value=112.)
true_params.add('phase',value=1.5)
true_params.add('mean',value=14.)

sample_times = np.random.uniform(0.,365.,size=ndata)
mags = light_curve_model(true_params,sample_times)

mag_err = np.random.uniform(0.1,0.3,size=ndata)
mags += np.random.normal(0., mag_err)

outlier_idx = np.random.uniform(size=ndata) < outlier_fraction
mags[outlier_idx] += np.random.normal(0.,np.random.uniform(3.,5., size=np.sum(outlier_idx)))

plt.errorbar(sample_times,mags,mag_err,marker='o',linestyle='none')
```

```python
def lightcurve_guess(x,y):
    params = lmfit.Parameters()
    params.add('amp',value=1,min=1,max=2)
    params.add('period',value=150,min=100,max=200)
    params.add('phase',value=1,min=0,max=2*np.pi)
    params.add('mean',value=13,min=12,max=16)
    return params
```

```python
mcfit = kid_readout.analysis.mcfit.MCMCFitter(x_data=sample_times,
                                              y_data=mags,
                                              errors=mag_err,
                                              model=light_curve_model,
                                              guess=lightcurve_guess)
```

```python
mcfit.setup_sampler()
```
