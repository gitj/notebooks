```python
zbd_voltage = np.array([nm.zbd_voltage for nm in powsweep])
zbd_power = zbd_voltage/2000
f_0 = np.array([nm.fit_params['f_0'].value for nm in powsweep])
f0b = np.array([pkl['noise_off_sweeps'][2]['f_0'].value for pkl in pkls])
qi = np.array([nm.Q_i for nm in powsweep])
q = np.array([nm.fit_params['Q'].value for nm in powsweep])
frac_f0 =(f_0/f_0.max()-1)
load_temp = np.array([nm.sweep_secondary_load_temperature for nm in powsweep])
```

```python
plot(load_temp,frac_f0*1e6,'o-')
plot(load_temp,(f0b/f_0.max()-1)*1e6,'x-',mew=2)
grid()
ylim(-170,0)
```

```python
pp = np.polyfit(frac_f0,zbd_power,2)
print pp
plot(zbd_power*1e6,1e6*frac_f0,'o',label='data')
plot(np.polyval(pp,frac_f0[:-1])*1e6,1e6*frac_f0[:-1],label='quadratic fit')
grid()
legend(loc='upper right', prop=dict(size='small'))
xlabel('Measured mm-wave power at source ($\mu$W)')
ylabel('Fractional frequency shift (ppm)')
#ylim(-160,0)
```

```python
loglog(zbd_power,-frac_f0,'.-')
grid()
xlabel('Measured mm-wave power at source (W)')
ylabel('Fractional frequency shift')
```

```python
plot(zbd_power*1e6,1e6/qi,'o',label='Qi')
plot(zbd_power*1e6,1e6/q,'o',label='Qr')
grid()
legend(loc='upper left', prop=dict(size='small'))
ylabel('$10^6/Q$')
xlabel('Measured mm-wave power at source ($\mu$W)')
```
