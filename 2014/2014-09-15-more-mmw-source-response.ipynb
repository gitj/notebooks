## First, a look at the Q's

```python
plot([mmw.resonator.f_0 for mmw in mmws],[mmw.resonator.Q for mmw in mmws],'.',label='Q')
plot([mmw.resonator.f_0 for mmw in mmws],[mmw.resonator.Q_i for mmw in mmws],'o',label='Qi')
xlabel('Resonator freq [MHz]')
ylabel('quality factor')
legend(loc='upper right',prop=dict(size='small'))
```

## The response of all detectors

```python
figure(figsize=(16,6))
for mmw in mmws[:-2]:
    plot(mmw.mmw_freq/1e9,1e6*(mmw.mmw_frac_response))
title('Response of all detectors')
ylabel('Frac. freq response [ppm]')
xlabel('mm-wave source freq [GHz]')
ylim(0,6)
```

# The relative response of the detectors versus a single detector

```python
figure(figsize=(10,6))
ref = mmws[3]
for mmw in mmws[:-3]:
    plot(mmw.mmw_freq/1e9,(mmw.mmw_frac_response/ref.mmw_frac_response))
#ylim(-70,-50)
ylim(0,3)
title('Relative response of all detectors versus #3')
ylabel('response / (det. #3 response)')
xlabel('mm-wave source freq [GHz]')
```

## Detailed response of each detector

In the following plots, the right hand plot shows the data. The yellow circle shows the model fit to the measured resonance sweep (black x's). The blue cross shows the 'zero-point' which is the readout value when sitting on the resonator with no change in the load. If you look closely, you can see a small swarm of green dots near it, these are the points measured when the mm-wave source was off. The red points are all of the points measured when the mm-wave source was on.
