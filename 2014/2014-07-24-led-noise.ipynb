```python
dd = df[(df.resonator_id==2) & (df.f_0_err/df.f_0 < 1e-7) & (df.f_0_err/df.f_0 > 1e-8)]
plot(dd.f_0,'.-')
```

```python
plt.rcParams['font.size']=16
figure(figsize=(8,8))
for k in range(len(dd)):
    loglog(dd.pca_freq.iat[k],dd.pca_eigvals.iat[k][1,:],color = plt.cm.spectral(dd.pca_eigvals.iat[k][1,20:80].mean()/5e-17))
    loglog(dd.pca_freq.iat[k],dd.pca_eigvals.iat[k][0,:],color = plt.cm.spectral(dd.pca_eigvals.iat[k][0,20:80].mean()/5e-17))
ylim(5e-19,1e-16)
xlim(10,3e4)
xlabel('Hz')
ylabel('1/Hz')
title('Noise spectra with increasing LED illumination')
grid()
savefig('/home/gjones/paper_plots/2014-07-24-led-noise-spectra.pdf',bbox_inches='tight')
```
