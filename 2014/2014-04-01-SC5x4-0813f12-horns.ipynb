```python
figure(figsize=(4,4))
betas = []
to_plot = [0,3,5,12]
plt.xticks(rotation=45,size=14)
plt.yticks(size=14)
colors = plt.cm.spectral(np.linspace(0.1,0.9,7))
symbols = ['o','d','^','v']
xs = np.linspace(-0.0008,0,100)
pltidx = 0
for idx in range(20):
    deltaf = (df[df.ridx==idx]['f0s']-fbase[idx])/fbase[idx]
    deltaq = 1/df[df.ridx==idx].Qi
    msk = ((deltaf > -0.0005) & (deltaf <= 0) &
           (deltaq >= 0) & (deltaq < 0.0003))
    if msk.sum() < 1:
        betas.append(np.nan)
        continue
    deltaf = deltaf[msk]
    deltaq = deltaq[msk]
    pp = np.polyfit(2*deltaf,deltaq,1)
    if idx in to_plot:
        plot(xs,np.polyval(pp,xs),color=colors[pltidx])
        plot(2*deltaf,deltaq,symbols[pltidx],color=colors[pltidx])
        pltidx += 1
        
    print idx,1/pp[0]
    betas.append(abs(1/pp[0]))
betas = np.array(betas)
xlim(-0.0006,0)
ylim(0,0.000015)
xlabel('$2\Delta f / f$')
ylabel('$1/Q_i$')
savefig('/home/gjones/paper_plots/sc5x4_0813f12_al_horns_deltaf_qi_optical.pdf',bbox_inches='tight')
```

```python
from kid_readout.analysis import kid_eqns
```

```python
def beta(Tbase,Tc,fread):
    S1,S2 = kid_eqns.s1s2(Tbase,Tc,fread)
    return S2/S1
```

```python
ff = np.linspace(80,170,100)
figure(figsize=(4,4))
plot(fbase[fbase<160],betas[fbase<160],'o')
plot(ff,beta(0.35,1.45,ff*1e6))
ax = gca()
ax.xaxis.set_major_locator(plt.MaxNLocator(5))
ylim(0,90)
xlabel('$f_{res}$ [MHz]')
ylabel('$\\beta$')
savefig('/home/gjones/paper_plots/sc5x4_0813f12_al_horns_optical_beta.pdf',bbox_inches='tight')
```
