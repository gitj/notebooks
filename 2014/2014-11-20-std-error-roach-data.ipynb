```python
x = np.zeros((4096,))
x[0] = 1
```

```python
def calc_mean_std(x):
    xf = kid_readout.utils.fftfilt.fftfilt(kid_readout.utils.data_block.lpf,x)[256:]
    std = xf.std()/np.sqrt(len(xf)/256.)
    avg = xf.mean()
    return avg,std
```

```python
gold_mean = []
gold_std = []
calc_mean = []
calc_std = []
for k in range(1000):
    x = np.random.randn(4096)+np.random.randn(4096)*1j
    gold_mean.append(x.mean())
    gold_std.append(x.std())
    avg,std = calc_mean_std(x)
    calc_mean.append(avg)
    calc_std.append(std)
```

```python
gold_mean = np.array(gold_mean)
gold_std = np.array(gold_std)
calc_mean = np.array(calc_mean)
calc_std = np.array(calc_std)
```
