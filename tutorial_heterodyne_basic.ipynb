Create a RoachHeterodyne class instance

```python
ri = roach_interface.RoachHeterodyne()
```

Initialize the ROACH and sampling clock:

```python
ri.initialize()
```

Set ADC and DAC attenuators so we don't blast huge signals

```python
ri.set_dac_attenuator(55)
```

```python
ri.set_adc_attenuator(27)
```

Now let's start with generating a single tone. The final output frequency will be $F_{LO}+45.6789$ MHz

`nsamp` indicates the number of samples used in the playback buffer which sets the frequency resolution

The actual frequency that the tone is at (after quantizing to units of the frequency resolution implied by `nsamp` is returned.

```python
ri.set_tone_freqs(np.array([45.6789]),nsamp=2**20)
```

Assuming the system is connected up such that there is a path from the DAC to the ADC, we can now view the tone. First, grab some raw ADC samples. The following command will provide 4096 samples of both ADCs, stored in `x` and `y` here.

```python
x,y = ri.get_raw_adc()
```

Now take a look and observe that we have a sine and a cosine as expected

```python
plot(x[:100])
plot(y[:100])
```

We can also view a spectrum of this data using the matplotlib/pylab `psd` function. Note that psd is smart enough to properly deal with both real and complex waveforms. Thus we can view the power spectrum of a single ADC like this:

```python
pxx,fr = psd(x,NFFT=1024,Fs=512.0) # NFFT specifies frequency resolution,
```

Or we can view the overall power spectrum of the complex signal like this:

```python
pxx,fr = psd(x+1j*y,NFFT=1024,Fs=512.0)
```

Note that there is an image component at -45 MHz which is ~15 dB down from the desired signal. This will vary depending on how the ADC ends up initialized, something that needs to be fixed.

Try playing around with set_adc_attenuator, and get_raw_adc to get a feel for how the attenuator value affects the ADC level and waveform. Some things to keep in mind:

 * The ADC counts are proportional to *voltage*, while the attenuator setting is in dB *power*
 * psd plots power, as you'd expect
 * The ADC has 12 bits of resolution, meaning 4096 distinct values can be represented, which are the integers from -2048 to 2047 inclusive.

We can also look at the waveform that we told the DAC to produce. The shoulders on the tone you see below are because of the Hann window used by default by psd.

```python
psd(ri.i_wave+1j*ri.q_wave,NFFT=4096,Fs=512.0)
```

Now let's grab a bit of data from the FFT bin. Initially we're going to turn off demodulation to look at the raw data.

```python
d,addr = ri.get_data(2,demod=False)
```

```python
d.shape # take a look at how the data is organized
```

We have 8192 samples from the single FFT bin, hence the shape is 8192,1. Each sample is a complex number. The `addr` array can be used to check for any dropped data. We won't worry about that here.

Let's have a look at the time domain representation of the data:

```python
plot(d[:50,0].real)
plot(d[:50,0].imag)
```

In the complex plane:

```python
plot(d.real,d.imag,'.')
axis('equal')
grid()
```

In the power spectrum: Make sure you understand where Fs comes from

```python
pxx,fr = psd(d[:,0],NFFT=1024,Fs=512.0/ri.nfft)
```

At what frequency did the peak occur?

```python
fr[pxx.argmax()]
```
