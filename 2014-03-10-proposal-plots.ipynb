```python
arch = kid_readout.analysis.noise_archive.load_archive('/home/data/archive/StarCryo_3x3_0813f5_HPD_Dark_1.npy')
arch['round_temp'] = np.round(arch.end_temp*1000/10)*10
```

```python
arch = arch[(arch.power_dbm > -114) & (arch.power_dbm <= -98)]
arch = arch[np.abs(arch.end_temp -0.235) > 0.01]
arch = arch[(arch.Q_i > 0) & (arch.Q_i < 4e5)]
idxs = np.unique(arch['ridx'])
arch = arch.sort(['end_temp'])
```

```python
fig = plt.figure()
ax2 = fig.add_subplot(111)
ax = ax2.twinx()
for rnum in [0]:
    this = arch[(abs(arch.power_dbm + 107) < 1) &
                (arch.ridx == rnum) &
                (arch.round_temp != 110) &
                (arch.round_temp != 160)]
    frac_f0 = this.f_0/this.f_0.max()
    temp = this.end_temp*1000
    ax.plot(temp,(frac_f0-1)*100,'.-',color='r')
    ax2.plot(temp,this.Q_i,'.-',color='k')
    #ax2.plot(this.end_temp*1000,this.Q,'.-',color='cyan')
ax.set_ylim(-.1,.002)
ax2.set_ylim(0,3e5)
ax.set_xlim(40,440)
ax.set_ylabel('Fractional frequency shift [%]')
ax2.set_ylabel('Internal quality factor')
ax2.set_xlabel('Physical temperature [mK]')
ax.annotate("",xytext=(temp.iat[-16]+10,(frac_f0.iat[-16]-1)*100),
            xy=(440,(frac_f0.iat[-16]-1)*100), xycoords='data',
            textcoords='data',
            arrowprops=dict(arrowstyle='fancy',shrinkA=5,shrinkB=5,fc="w",ec="r")
            )
ax2.annotate("",xytext=(temp.iat[6]-10,this.Q_i.iat[6]),
            xy=(40,this.Q_i.iat[6]), xycoords='data',
            textcoords='data',
            arrowprops=dict(arrowstyle='fancy',shrinkA=5,shrinkB=5,fc="w",ec="k")
            )
fig.savefig('/home/gjones/lekid_frequency_shift.pdf',bbox_inches='tight')            
```

```python
def smooth(x,n=5):
    return np.convolve(np.ones((n,))/float(n),x,mode='same')
```

```python
fig = plt.figure()#figsize=(6,4))
data = np.zeros((spectra.values()[0][0][0].shape[0],6))
n = 0
for k in [1,6,8,11,13,17]:
    ss = spectra.values()[k]
    for (fr,pxx) in ss[:1]:
        loglog(smooth(fr),smooth(pxx),alpha=0.7)
        data[:,n] = pxx
    n+=1
xlim(0.05,2e4)
axhline(35,linestyle='--',lw=1,color='k')
ylim(10,300)
ylabel('NET [$\mu$K$\sqrt{s}$]')
xlabel('Frequency [Hz]')
text(0.1,30,"35 $\mu$K$\sqrt{s}$",va='top')
fig.savefig('/home/gjones/lekid_many_noise_spectra.pdf',bbox_inches='tight')
#axhspan(27,60,alpha=0.2)
```
