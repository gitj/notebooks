```python
import scipy.integrate
```

First, define the blackbody curve:

$$ B_{\nu}(T) = \frac{h\nu^3}{c^2 (e^{h\nu/kT}-1)} n_{modes}\lambda^2 $$

Where I'm using
$$ A\Omega = n_{modes}\lambda^2 $$

Since we have the low pass filter in place, we'll assume $n_{modes} = 1$ giving:

$$ B_{\nu}(T) = \frac{h\nu}{e^{h\nu/kT}-1} $$


```python
def Bnu(nu,T,emissivity=.75):
    return emissivity*6.626e-34*nu/(np.exp((6.626e-34/1.38e-23)*(nu/T))-1)
```

Now integrate over the band, from 120 GHz to 170 GHz, to get power as a function of temperature

```python
def P(T):
    def _P(T):
        return scipy.integrate.quad(Bnu,120e9,170e9,args=(T))[0]
    return np.vectorize(_P)(T)
```

```python
T = np.linspace(1,10,200)
```

```python
figure(figsize=(12,6))
subplot(121)
plot(T,P(T)*1e12)
title("power vs. temperature")
ylabel('Power (pW)')
xlabel('Load Temperature (K)')
grid()
subplot(122)
plot(T[:-1],(np.diff(P(T))/np.diff(T)))
title("dP/dT")
ylabel('dP/dT')
xlabel('Load Temperature (K)')
grid()
```

$$ Q_{\nu}(T) = \frac{1}{e^{h\nu/kT}-1} $$

```python
def Qnu(nu,T,emissivity=.75):
    return emissivity/(np.exp((6.626e-34/1.38e-23)*(nu/T))-1)
```

```python
def Q(T):
    def _Q(T):
        return scipy.integrate.quad(Qnu,120e9,170e9,args=(T))[0]
    return np.vectorize(_Q)(T)
```

```python
figure(figsize=(12,6))
subplot(121)
plot(T,Q(T))
title("Q(T)")
ylabel('photons/second')
xlabel('Load Temperature (K)')
grid()
subplot(122)
plot(T[:-1],(np.diff(Q(T))/np.diff(T)))
title("dQ/dT")
ylabel('dQ/dT')
xlabel('Load Temperature (K)')
grid()
```

$$n_{qp}=n^* \sqrt{1+2\Gamma_o \tau_{max}/N^*}-n^*$$

$$n_{qp}=n^* \sqrt{1+\frac{4P_o \tau_{max}}{(h\nu) V n^*}}-n^*$$

```python
def nqp_optical(P,nstar=100,Vum=1800,tau_max=500e-6,nu=150e9):
    Gamma = 2*P/(6.626e-34*nu)
    nqp = nstar*(np.sqrt(1+2*Gamma*tau_max/(nstar*Vum))-1)
    return nqp
```

```python
plot(T,nqp_optical(P(T)),label='$n^*$=100, $\\tau_{max}$=500')
plot(T,nqp_optical(P(T),tau_max=125e-6),label='$n^*$=100, $\\tau_{max}$=125')
plot(T,nqp_optical(P(T),nstar=50),label='$n^*$=50,  $\\tau_{max}$=500')
title("optical quasiparticle density")
ylabel('$n_{qp}$ [$\mu m^{-3}$]')
xlabel('Load Temperature (K)')
legend(loc='upper left',prop=dict(size='small'))
grid()
```

```python
N0 = 1.72e10
Delta = kid_eqns.kBeV*1.74*1.46
def deltaf(nqp,alpha=0.66,S2=3.8):
    return alpha*S2*nqp/(4*N0*Delta)
```

```python
def Qi_nqp(nqp,alpha=0.66,S1=0.121):
    return 2*N0*Delta/(alpha*S1*nqp)
```
