```python
f1 = figure()
ax1 = f1.add_subplot(111)
f2 = figure(figsize=(8,12))
ax2 = f2.add_subplot(311)
ax3 = f2.add_subplot(312)
ax4 = f2.add_subplot(313)
det = 0

for name,tsg in nc.timestreams.groups.items():
    epoch = tsg.variables['epoch'][0]
    idx = np.flatnonzero(nc.hw_state.epoch<epoch)[-1]
    atten = nc.hw_state.dac_atten[idx]
    if atten != 18.0:
        continue
    print name
    d = tsg.variables['data'][det,:].view('complex128')
    dmean = d.mean()
    ax2.plot([nc.cryo.bridge_temp_value[idx]],[10*np.log10(np.abs(dmean)**2)],'bo')
    pxx,fr = mlab.psd(d-dmean,NFFT=2**18,Fs=256e6/2**14)
    msk = fr>0
    ax3.plot([nc.cryo.bridge_temp_value[idx]],[10*np.log10(pxx[msk][0])],'bo')
    ax4.plot([nc.cryo.bridge_temp_value[idx]],[(pxx[msk][-1000:].mean())],'ro')

    ax1.semilogx(fr[msk],10*np.log10(pxx[msk]))
ax2.grid()
ax3.grid()
ax4.grid()
ax2.set_ylabel('relative transmission (dB)')
ax3.set_ylabel('red noise power (dB)')
ax4.set_ylabel('white (electronic) noise power (linear)')
ax4.set_xlabel('Temperature (K)')
ax2.set_title('Star Cryo 3x3 chip')
title('%s %.1f' % (name,atten))

```
