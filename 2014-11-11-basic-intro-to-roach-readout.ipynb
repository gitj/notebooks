If you want to use loopback mode, enable it as follows.

**Note:** In loopback mode, the waveforms will be full scale (no attenuation) so the filterbank may overflow a bit or otherwise act funny, so don't be alarmed if the results are not numerically perfect.

```python
ri.r.write_int('loopback',1)  # this is a direct katcp write to the loopback register
```

Now let's look at some data from the filterbank. Initially, we will turn off the demodulation (which happens in the python code after the data is received

```python
d,addr = ri.get_data_katcp(16,demod=False)
```

Here we have read 4 channels at a time, each of which represents a bandwidth of 512 MHz / (2 * nfft) = $512\times10^6/2^{15}$ = 15.625 kSample/s, so all four of them together yeild the quoted ~62.5 kSample/s

The resulting data has one column per tone read out

```python
d.shape
```

Here is a quick look at the power spectrum, not demodulated

```python
blah = psd(d[:,0],NFFT=1024,Fs=ri.fs*1e6/(2*ri.nfft))
```

The center frequencies of the filterbank channels are given by:

```python
ri.fs*ri.fft_bins[0]/2.**15
```

So we can plot the actual frequencies of the filterbank output as follows

```python
f0s = ri.fs*1e6*ri.fft_bins[0]/2.**15
pxx,fr = mlab.psd(d[:,0],NFFT=1024,Fs=ri.fs*1e6/(2*ri.nfft))
plot(fr+f0s[0],10*np.log10(pxx))
ylabel('dB')
xlabel('Hz')
grid()
```
