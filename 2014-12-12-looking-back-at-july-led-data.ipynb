```python
r0['frac_f0'] = 1-r0.f_0/r0.f_0.max()
```

```python
def dxdP(Po,taumax=100e-6,nstar=400,alpha=0.66,N0=1.72e10,Delta=1.76*1.38e-23*1.46,Vl=1870,nu=150e9,eta=1,etapb=0.7,
         S2=3.8):
    a1 = alpha*S2*taumax*etapb*eta/(2*N0*Delta*6.626e-34*nu*Vl/1.6e-19)
    a2 = np.sqrt(1+4*etapb*eta*Po*taumax/(6.626e-34*nu*Vl*nstar))
    return a1/a2
```

```python
loglog(((r0.frac_f0**2*(6.626e-34*3e8/1550e-9)/(r0.noise_fit_device_noise))),
       1e-12*(r0.noise_fit_device_noise)/(r0.frac_f0*6.626e-34*3e8/1550e-9),'.')
po = np.linspace(1e-14,1e-12,1000)
loglog(po,1e-12*dxdP(po),label='taumax=100, nstar=400, alpha=0.66')
ylim(1e-5,1e-4)
loglog(po,1e-12*dxdP(po,alpha=0.26),label='taumax=100, nstar=400, alpha=0.26')
legend(loc='lower left',prop=dict(size='small'))
title('1550 nm LED')
xlabel('$P_{abs} [W]$')
ylabel('$dx/dP$ [1/pW]')
       
```

```python
def Sxx(x,responsivity,Sxx0,bandwidth,frequency=150e9):
    return x*responsivity*6.626e-34*frequency +Sxx0+ x**2/bandwidth
```
