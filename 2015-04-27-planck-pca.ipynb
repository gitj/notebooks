```python
def plot_planck(data,shape=(100,99),nrows=None,ncols=None,figsize=(12,12)):
    if nrows is None:
        if data.shape[0] <=4:
            nrows,ncols = 2,2
        else:
            nrows,ncols = 3,3
    fig,axs = plt.subplots(nrows,ncols,figsize=figsize)
    for k in range(data.shape[0]):
        ax = axs.flatten()[k]
        ax.imshow(data[k].reshape(shape),aspect='auto',interpolation='nearest',cmap=cm.cubehelix)
```

```python
plot_planck(red)
```

```python
plot_planck(smplk.reshape(9,-1))
```
