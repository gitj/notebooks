```python
for resid in [0,1]:
    r1 = df1[df1.resonator_id==resid]
    r2 = df2[(df2.resonator_id==resid) & (df2.timestream_modulation_duty_cycle==1)]
    figure()
    plot(r1.sweep_primary_load_temperature,r1.f_0,'.')
    plot(r2.sweep_primary_package_temperature,r2.f_0,'x',mew=2)
    if resid==1:
        x=np.linspace(0,80,1000)
        plot(x,158/np.sqrt(1+(x/98.)**4))
    if resid==0:
        x=np.linspace(0,80,1000)
        plot(x,201.5/np.sqrt(1+(x/80.)**4))
        ylim(200.5,202)
    #ylim(100,2e4)
    xscale('log')
    yscale('log')
```

```python
for k,row in df1[df1.resonator_id==1].iterrows():
    if row.f_0 < (158/np.sqrt(1+(row.sweep_primary_load_temperature/98.)**4)):
        color='k'
        lw = 1
    else:
        color = 'r'
        lw=2
    plot(row.sweep_model_freqs_MHz,20*np.log10(abs(row.sweep_model_normalized_s21)),color=color,alpha=.5,lw=lw)
```
